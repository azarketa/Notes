# ON THE RADIAL LAPLACE EQUATION

**THE GOVERNING PDE**

We start with the equation written in the figure:

$$
\frac{1}{r}\frac{\partial}{\partial r}\left(r \frac{\partial v_x}{\partial r}\right) + \frac{\partial^2 v_x}{\partial x^2} = 0.
$$

This is a **Helmholtz-type equation** in cylindrical coordinates, but with no (\theta)-dependence. Physically:

* $v_x(r,x)$ is the **axial velocity perturbation** downstream of the actuator disk.
* The first term
  $\tfrac{1}{r},\partial_r ( r,\partial_r v_x )$
  accounts for the **radial diffusion/variation** of the axial velocity.
* The second term
  $\partial_x^2 v_x$
  captures how the disturbance decays (or oscillates) in the axial direction.

This PDE comes directly from applying linearized continuity and momentum balances to the incompressible, axisymmetric flow around the disk. It essentially says: the axial disturbance spreads radially and decays axially in a separable way.

::::{admonition} Deriving the axisymmetric Laplace equation for $v_x(r,x)$
:class: note

We consider steady, incompressible, axisymmetric flow in cylindrical coordinates $((r,\theta,x))$ with no $\theta$ dependence. The goal is to show

$$
\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial v_x}{\partial r}\right)
+\frac{\partial^2 v_x}{\partial x^2}=0.
$$

* Assumptions:

  * Incompressible: $\nabla\cdot\mathbf{v}=0$.
  * Axisymmetric: $\partial/\partial\theta=0$.
  * Steady: $\partial/\partial t=0$.
  * High-Re (viscous terms neglected in the near-field modal problem).
  * Small perturbations around a base flow (used in Route 2).

---

**ROUTE 1 — POTENTIAL-FLOW (SHORTEST PATH)**

Assume the meridional flow is locally irrotational, so there exists a potential $\phi(r,x)$ with

$$
\mathbf{v}=\nabla\phi,\qquad
v_r=\frac{\partial\phi}{\partial r},\qquad
v_x=\frac{\partial\phi}{\partial x}.
$$

* **STEP 1 — INCOMPRESSIBILITY IN CYLINDRICAL COORDINATES**

    For axisymmetric flow ($\partial/\partial\theta=0$), incompressibility reads
    
    $$
    \frac{1}{r}\frac{\partial}{\partial r}(r v_r)+\frac{\partial v_x}{\partial x}=0.
    $$
    
    With $v_r=\partial\phi/\partial r$ and $v_x=\partial\phi/\partial x$, this is equivalent to
    
    $$
    \nabla^2\phi=\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial \phi}{\partial r}\right)
    +\frac{\partial^2 \phi}{\partial x^2}=0.
    $$


* **STEP 2 — DIFFERENTIATE LAPLACE'S EQUATION**

    Differentiate this with respect to $x$ and use $v_x=\partial\phi/\partial x$:
    
    $$
    \frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial v_x}{\partial r}\right)
    +\frac{\partial^2 v_x}{\partial x^2}=0.
    $$
    
    This directly yields the governing PDE.

---

**ROUTE 2 — LINEARIZED, STEADY, INVISCID EULER**

* Start from the incompressible Euler equations (steady):

    $$
    \nabla\cdot\mathbf{v}=0,\qquad
    (\mathbf{v}\cdot\nabla)\mathbf{v}=-\frac{1}{\rho}\nabla p.
    $$

* **STEP 1 — CONTINUITY IN CYLINDRICAL COORDINATES**

    The incompressibility condition is
    
    $$
    \frac{1}{r}\frac{\partial}{\partial r}(r v_r)
    +\frac{1}{r}\frac{\partial v_\theta}{\partial \theta}
    +\frac{\partial v_x}{\partial x}=0.
    $$

    Axisymmetry removes the azimuthal derivative:
    
    $$
    \frac{1}{r}\frac{\partial}{\partial r}(r v_r)+\frac{\partial v_x}{\partial x}=0.
    $$
    
    This is the **radial continuity relation**.

* **STEP 2 — MOMENTUM EQUATIONS (FULL AXISYMMETRIC FORM)**

    * Radial:
 
      $$
      v_r\frac{\partial v_r}{\partial r}
      +v_x\frac{\partial v_r}{\partial x}
      -\frac{v_\theta^2}{r}
      = -\frac{1}{\rho}\frac{\partial p}{\partial r}.
      $$
    
    * Azimuthal:
 
      $$
      v_r\frac{\partial v_\theta}{\partial r}
      +v_x\frac{\partial v_\theta}{\partial x}
      +\frac{v_r v_\theta}{r}=0.
      $$
    
    * Axial:
 
      $$
      v_r\frac{\partial v_x}{\partial r}
      +v_x\frac{\partial v_x}{\partial x}
      = -\frac{1}{\rho}\frac{\partial p}{\partial x}.
      $$

* **STEP 3 — BASE-PERTURBATION SPLIT**

    Take the base flow as
    
    $$
    \mathbf{V_0}=(0,V_{\theta0}(r),U_0),
    $$
    
    with perturbations
    
    $$
    \mathbf{v}'=(v_r,v_\theta',v_x),\qquad p=p_0+p'.
    $$
    
    Neglect quadratic perturbation terms.

* **STEP4  — LINEARIZED MOMENTUM EQUATIONS**

    * **Radial (linearized):**
    
        $$
        U_0\frac{\partial v_r}{\partial x}
        -\frac{2 V_{\theta0}}{r},v_\theta'
        = -\frac{1}{\rho}\frac{\partial p'}{\partial r}.
        $$
        
        * **Azimuthal (linearized, for reference):**
        
        $$
        U_0\frac{\partial v_\theta'}{\partial x}
        +v_r\frac{dV_{\theta0}}{dr}
        +\frac{V_{\theta0}}{r}v_r=0.
        $$
        
        * **Axial (linearized):**
        
        $$
        U_0\frac{\partial v_x}{\partial x}
        = -\frac{1}{\rho}\frac{\partial p'}{\partial x}.
        $$

* **STEP 5 — SIMPLIFIED NEAR-FIELD FORM**

    If $v_\theta'$ is neglected for the decay analysis away from the disk, the radial and axial momentum reduce to
    
    $$
    U_0\frac{\partial v_r}{\partial x}
    = -\frac{1}{\rho}\frac{\partial p'}{\partial r},
    \qquad
    U_0\frac{\partial v_x}{\partial x}
    = -\frac{1}{\rho}\frac{\partial p'}{\partial x}.
    $$
    
    Differentiate the first with respect to $r$ and the second with respect to $x$, then subtract to eliminate $p'$:
    
    $$
    \frac{\partial^2 v_r}{\partial x\partial r}
    =\frac{\partial^2 v_x}{\partial x^2}.
    $$

* **STEP 6 — COMBINE WITH CONTINUITY**

Take $\partial/\partial x$ of the continuity equation:

$$
\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial v_r}{\partial x}\right)
+\frac{\partial^2 v_x}{\partial x^2}=0.
$$

Now use $\partial v_r/\partial x=\partial v_x/\partial r$ to obtain

$$
\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial v_x}{\partial r}\right)
+\frac{\partial^2 v_x}{\partial x^2}=0.
$$

---

**SUMMARY**

* **Route 1 (potential flow)** derives the PDE directly from Laplace’s equation for the potential.
* **Route 2 (linearized Euler)** shows explicitly how the incompressibility condition in cylindrical coordinates, together with the linearized momentum balances, lead step by step to the same governing PDE.

::::

---

**SEPARATION ON VARIABLES**

The **method of separation of variables** provides a systematic way to construct solutions, and the resulting **characteristic equation** determines the admissible modes. We assume the solution can be written as a product of a purely radial and a purely axial function:

$$
v_x(r,x) = R(r)X(x).
$$

Substitute this into the PDE:

$$
\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{dR}{dr}X(x)\right)
+\frac{d^2 X}{dx^2}R(r)=0.
$$

Divide through by $R(r)X(x)$ (assuming neither is identically zero):

$$
\frac{1}{R(r)}\frac{1}{r}\frac{d}{dr}\left(r\frac{dR}{dr}\right)
+\frac{1}{X(x)}\frac{d^2 X}{dx^2}=0.
$$

At this point, the left-hand side is written as a sum of two terms, one depending only on $r$, the other only on $x$. The only way this can hold for all $(r,x)$ is if both terms equal the same constant, say $-\lambda^2$. This gives us two ordinary differential equations (ODEs):

* Radial ODE:
  $$
  \frac{1}{r}\frac{d}{dr}\left(r\frac{dR}{dr}\right) + \lambda^2 R = 0.
  $$

* Axial ODE:
  $$
  \frac{d^2 X}{dx^2} - \lambda^2 X = 0.
  $$

The parameter $\lambda$ is called the **separation constant**, and physically represents a radial wavenumber.

::::{admonition} The constant $\lambda^{2}$ in the separation of variables step
:class: note

We can write the generic equation derived from the separation of variables condition above as follows:

$$
\frac{1}{R(r)}\frac{1}{r}\frac{d}{dr}\left(r\frac{dR}{dr}\right) = C,
\qquad
\frac{1}{X(x)}\frac{d^2 X}{dx^2} = -C.
$$

---

**WHY CHOOSE $C=-\lambda^2$?**

We are free to denote the separation constant in any way. But the **sign matters** for the type of solutions we get. There are three possibilities:

1. **$C=+\lambda^2$** with $\lambda$ real

   * Axial ODE: $X'' + \lambda^2 X = 0$ → sinusoidal solutions $\cos(\lambda x)$, $\sin(\lambda x)$.
   * Radial ODE: $\frac{1}{r}(rR')' - \lambda^2 R=0$, which corresponds to the modified Bessel equation, solutions $I_0(\lambda r), K_0(\lambda r)$ (exponential in $r$).
   * Physically: not desirable here, because we want *oscillations* radially (captured by ordinary Bessel functions) and *decay* axially, not the other way around.

2. **$C=-\lambda^2$** with $\lambda$ real

   * Axial ODE: $X'' - \lambda^2 X=0$ → exponential solutions $e^{\pm \lambda x}$. This gives the desired *decay* away from the disk.
   * Radial ODE: $\frac{1}{r}(rR')' + \lambda^2 R=0$ → Bessel’s equation of order zero, with oscillatory $J_0(\lambda r)$ and $Y_0(\lambda r)$.
   * Physically: exactly the structure we want — radial oscillations shaped by Bessel functions, and exponential decay in $x$.

3. **$C=0$**

   * Axial ODE: $X''=0$ → linear solutions in $x$.
   * Radial ODE: $\frac{1}{r}(rR')'=0$ → $R=A\ln r + B$.
   * Physically: gives only trivial or unbounded modes, not useful for actuator-disk problems.

So by choosing $C=-\lambda^2$, we force the mathematics to produce **Bessel functions radially** and **exponentially decaying modes axially** — the physically correct behavior in this geometry.

---

**WHY SQUARE $\lambda$?**

The square is introduced for two reasons:

1. **Mathematical convenience**:

   Writing $C=-\lambda^2$ ensures that $\lambda$ comes out real and positive from the eigenvalue problem (zeros of Bessel functions). Otherwise we would constantly have to deal with $\pm\sqrt{-C}$, which is clumsy.

2. **Spectral interpretation**:

   The radial equation is a **Sturm–Liouville eigenvalue problem**:
   $$
   \frac{1}{r}\frac{d}{dr}\left(r\frac{dR}{dr}\right) + \lambda^2 R = 0.
   $$
   Such problems always yield a spectrum of **nonnegative eigenvalues $\lambda^2$**. The eigenvalues themselves ($\lambda$) are wavenumbers, while the squares appear naturally in the operator. Using $\lambda^2$ matches the standard Sturm–Liouville notation and interpretation.

---

**PHYSICAL INTERPRETATION OF $\lambda$**

* $\lambda$ has dimensions of **inverse length**, and plays the role of a radial wavenumber.
* Larger $\lambda$ means finer radial oscillations and faster axial decay ($e^{-\lambda x}$).
* The discrete set ${\lambda_n}$ comes from boundary conditions, and each corresponds to a physically realizable eigenmode of the flow.

---

**SUMMARY**

We set the separation constant as $-\lambda^2$ because that choice gives the **correct physical structure** (Bessel modes in $r$, exponential decay in $x$). The square ensures the eigenvalue problem is written in the standard Sturm–Liouville form, with a real, nonnegative spectrum $\lambda^2$.

::::

---

**SOLUTION OF THE AXIAL ODE**

The axial equation is simple:

$$
\frac{d^2 X}{dx^2} - \lambda^2 X = 0.
$$

Its general solution is:

$$
X(x) = A e^{-\lambda x} + B e^{+\lambda x}.
$$

To ensure boundedness for $x\to +\infty$ downstream of the disk, we retain only the decaying exponential:

$$
X(x) = A e^{-\lambda x}, \qquad x>0.
$$

By symmetry, or by matching at the disk plane, the upstream solution for $x<0$ is taken as

$$
X(x) = A e^{-\lambda |x|}.
$$

Thus, every radial mode is paired with an exponential decay away from the disk.

---

**SOLUTION OF THE RADIAL ODE**

The radial ODE is recognized as **Bessel’s equation of order zero**:

$$
\frac{1}{r}\frac{d}{dr}\left(r\frac{dR}{dr}\right) + \lambda^2 R = 0.
$$

The general solution is a linear combination of Bessel functions:

$$
R(r) = J_0(\lambda r) + bY_0(\lambda r).
$$

* $J_0$ is the Bessel function of the first kind (regular at $r=0$).
* $Y_0$ is the Bessel function of the second kind (singular at $r=0$).

::::{admonition} Deriving the form of the radial ODE
:class: note

**REDUCTION TO THE CANONICAL BESSEL EQUATION**

Expand the radial operator and non-dimensionalize the independent variable:

$$
R''(r)+\frac{1}{r}R'(r)+\lambda^2 R(r)=0,
\qquad
\xi:=\lambda r,\quad R(r)=Y(\xi).
$$

Using $d/dr=\lambda,d/d\xi$ and $d^2/dr^2=\lambda^2,d^2/d\xi^2$,

$$
\lambda^2 Y''(\xi)+\frac{\lambda}{r}Y'(\xi)+\lambda^2 Y(\xi)=0
\Longrightarrow
Y''(\xi)+\frac{1}{\xi}Y'(\xi)+Y(\xi)=0.
$$

Multiplying by $\xi^2$ gives the **Bessel equation of order zero**:

$$
\xi^2 Y''+\xi Y'+\xi^2 Y=0.
$$

Thus the fundamental solutions are the order-zero Bessel functions:
$$
Y(\xi)=C_{1}J_0(\xi)+C_{2}Y_{0}(\xi),
\qquad
R(r)=C_{1}J_0(\lambda r)+C_{2}Y_{0}(\lambda r)
\quad
\Longrightarrow
\quad
R(r)=J_0(\lambda r)+bY_0(\lambda r)
\quad
;
\quad
b = C_{2}/C_{1}
$$

* **Solid disk (includes $r=0$):** $Y_0$ is singular at the axis $\Rightarrow b=0$.
* **Annulus ($R_i>0$):** both terms are admissible; $b$ is set by the two wall conditions at $r=R_i$ and $r=R$.

---

**FROBENIUS DERIVATION (WHY $J_{0}$ AND $Y_{0}$)**

Apply a power series $Y(\xi)=\sum_{k=0}^\infty a_k\xi^{k+s}$ about $\xi=0$. The indicial equation from the lowest power gives $s=0$ (double root). One linearly independent solution is analytic at the origin:

$$
J_0(\xi)=\sum_{m=0}^{\infty}\frac{(-1)^m}{(m)^2}\left(\frac{\xi}{2}\right)^{2m}
=1-\frac{\xi^2}{2^2}+\frac{\xi^4}{2^2 4^2}-\cdots .
$$

Because the indicial root is repeated, the second solution develops a logarithmic singularity; it is the Neumann function (Bessel of the second kind):

$$
Y_0(\xi)=\frac{2}{\pi}\left[\ln\left(\frac{\xi}{2}\right)+\gamma\right]J_0(\xi)
+\text{(analytic series in $\xi$)},
$$

with $\gamma$ the Euler–Mascheroni constant. Hence $Y_0(\xi)\to -\infty$ as $\xi\to 0^+$, while $J_0(\xi)$ is finite and $J_0(0)=1$.

**Conclusion near the axis:** if the physical domain includes $r=0$, regularity forces $C_2=0$; only $J_0$ is admissible there.

---

**CHARACTERISTIC EQUATIONS FROM BOUNDARY CONDITIONS**

Boundary conditions quantize $\lambda$ and fix $b$. The most common cases are:

* **Dirichlet at the outer wall, solid disk**

    If $R(0)$ finite and $R(R)=0$, then
    $$
    R(r)=J_0(\lambda r),\qquad J_0(\lambda R)=0
    \Longrightarrow
    \lambda_n=\frac{j_{0,n}}{R}.
    $$
    with $j_{0,n}$ the zeros of $J_0$.

<br/>
<br/>

* **Neumann at the outer wall, solid disk**

    If $R'(R)=0$ (no radial gradient at the wall), then
    $$
    -\lambda J_1(\lambda R)=0
    \Longrightarrow
    J_1(\lambda R)=0
    \Longrightarrow
    \lambda_n=\frac{j_{1,n}}{R},
    $$
    with $j_{1,n}$ the zeros of $J_1$.

<br/>
<br/>

* **Dirichlet–Dirichlet annulus ($R_i>0$)**

    Impose $R(R_i)=0$ and $R(R)=0$:
    $$
    \begin{aligned}
    &J_0(\lambda R_i)+bY_0(\lambda R_i)=0,\
    &J_0(\lambda R)+bY_0(\lambda R)=0.
    \end{aligned}
    $$
    For a nontrivial solution (some $b$), the determinant must vanish:
    $$
    J_0(\lambda R)Y_0(\lambda R_i)-J_0(\lambda R_i)Y_0(\lambda R)=0.
    $$
    Given a root $\lambda=\lambda_n$, one convenient formula is
    $$
    b_n=-\frac{J_0(\lambda_n R)}{Y_0(\lambda_n R)}
    = -\frac{J_0(\lambda_n R_i)}{Y_0(\lambda_n R_i)}.
    $$

<br/>
<br/>

* **Neumann–Neumann annulus**
    
    Impose $R'(R_i)=0$ and $R'(R)=0$:
    $$
    \begin{aligned}
    &J_1(\lambda R_i)+bY_1(\lambda R_i)=0,\
    &J_1(\lambda R)+bY_1(\lambda R)=0,
    \end{aligned}
    $$
    giving the characteristic equation
    $$
    J_1(\lambda R)Y_1(\lambda R_i)-J_1(\lambda R_i)Y_1(\lambda R)=0.
    $$

::::

---

**ROLE OF BOUNDARY CONDITIONS AND CHARACTERISTIC EQUATION**

As mentioned above, the **characteristic equation** arises from the radial boundary conditions. For an **annular disk** (hub with radius $R_i>0$) with Dirichlet-Dirichlet conditions, both $J_0$ and $Y_0$ are allowed. The boundary conditions are:
   $$
   R(R_i)=0,\qquad R(R)=0.
   $$
   This leads to the system
   $$
   J_0(\lambda R_i) + bY_0(\lambda R_i)=0, \qquad
   J_0(\lambda R) + bY_0(\lambda R)=0.
   $$
   For a nontrivial $b$, the determinant must vanish:
   $$
   J_0(\lambda R)Y_0(\lambda R_i)-J_0(\lambda R_i)Y_0(\lambda R)=0.
   $$
   This is the **characteristic equation** for $\lambda$. Its roots $\lambda_n$ are the admissible radial eigenvalues, and for each root the corresponding $b_n$ is fixed as
   :::{math}
   :label: eq_dirichlet_cond
   b_n = -\frac{J_0(\lambda_n R)}{Y_0(\lambda_n R)}.
   :::

---

**ASSEMBLING THE FULL SOLUTION**

Each admissible $\lambda_n$ defines a mode, with $F_n(r) = J_0(\lambda_n r) + b_n Y_0(\lambda_n r)$:

$$
v_x^{(n)}(r,x) = A_n e^{-\lambda_n |x|}F_n(r).
$$

The full solution is an expansion:

$$
v_x(r,x) = \sum_{n=1}^{\infty} A_n e^{-\lambda_n |x|}F_n(r).
$$

* The **radial dependence** is fixed by Bessel functions, shaped by the boundary conditions.
* The **axial dependence** is an exponential decay away from the disk plane.
* The **coefficients $A_n$** are determined by the disk physics: continuity of velocity and the Euler equation linking axial gradients, swirl, and the pressure drop.

::::{admonition} On the orthogonality condition, the $F_{n}(r)$ shape and the projection of velocity onto eigenfunctions
:class: note

**GENERAL STURM-LIOUVILLE STRUCTURE**

The separated radial equation is

$$
\frac{1}{r}\frac{d}{dr}\left(r \frac{dR}{dr}\right) + \lambda^2 R = 0,
\qquad R(R_i)=R(R)=0,
$$

where $R_i$ is the hub radius (possibly $R_i=0$).

This can be written in Sturm–Liouville form:

$$
\frac{d}{dr}\Big(p(r) \frac{dR}{dr}\Big) + \lambda^2 w(r)R = 0,
$$

with

* $p(r)=r$,
* $w(r)=r$.

Thus the problem is a **Sturm–Liouville eigenvalue problem** with weight $w(r)=r$ and eigenvalue $\lambda^2$.

---

**GENERAL ORTHOGONALITY RESULT**

If $R_m(r)$ and $R_n(r)$ are two eigenfunctions with eigenvalues $\lambda_m^2 \neq \lambda_n^2$, then

$$
\int_{R_i}^{R} w(r)R_m(r)R_n(r)dr = 0.
$$

Here $w(r)=r$. So the orthogonality condition is

$$
\int_{R_i}^{R} rR_m(r)R_n(r)dr = 0 \quad (m\neq n).
$$

This is the key property that allows us to project any admissible inlet distribution onto the eigenbasis.

---

**PROOF SKETCH OF ORTHOGONALITY**

Let $R_m$, $R_n$ be two solutions:

$$
\frac{1}{r}(r R_m')' + \lambda_m^2 R_m=0,
\qquad
\frac{1}{r}(r R_n')' + \lambda_n^2 R_n=0.
$$

Multiply the first equation by $R_{n}r$, the second by $R_{m}r$, subtract, and integrate from $R_i$ to $R$:

$$
\int_{R_i}^{R} \left[R_n (rR_m')' - R_m (rR_n')'\right]dr
+(\lambda_m^2-\lambda_n^2)\int_{R_i}^{R} r{}R_m{}R_n{}dr = 0.
$$

Integrate the first term by parts:

$$
\left[ r(R_m'R_n - R_n'R_m)\right]_{R_i}^{R}
+(\lambda_m^2-\lambda_n^2)\int_{R_i}^{R} r{}R_m{}R_n{}dr = 0.
$$

But under **Dirichlet–Dirichlet boundary conditions**, $R_m(R_i)=R_m(R)=R_n(R_i)=R_n(R)=0$, so the boundary term vanishes. Thus:

$$
(\lambda_m^2-\lambda_n^2)\int_{R_i}^{R} r{}R_m{}R_n{}dr = 0.
$$

Since $\lambda_m^2\neq \lambda_n^2$, we obtain:

$$
\int_{R_i}^{R} r{}R_m(r){}R_n(r){}dr = 0.
$$

This is the orthogonality condition.

:::{admonition} On the formal justification of orthogonality: self-adjointness
:class: tip

**WHERE ORTHOGONALITY REALLY COMES FROM**

The orthogonality of eigenfunctions is a **consequence of the Sturm–Liouville operator being self-adjoint** under a suitable inner product.

* The radial operator is

$$
\mathcal{L}[R] = -\frac{1}{r}\frac{d}{dr}\left(r \frac{dR}{dr}\right),
$$

with boundary conditions (e.g. Dirichlet at $R_i$ and $R$).

* The associated inner product is

$$
\langle f,g \rangle = \int_{R_i}^R r{}f(r){}g(r){}dr.
$$

* Self-adjointness means

$$
\langle f,\mathcal{L}[g] \rangle = \langle \mathcal{L}[f],{}g \rangle,
$$

for all admissible $f,g$ satisfying the boundary conditions.

From functional analysis, a self-adjoint Sturm–Liouville operator has:

1. **Real eigenvalues** $\lambda^2$.
2. **Orthogonal eigenfunctions** with respect to the weight function $w(r)=r$.

So the orthogonality is a direct consequence of the operator’s mathematical structure — not of a “clever manipulation.”

---

**WHY THE “MANIPULATION TRICK” WORKS**

When we multiply the $m$-equation by $R_n$, the $n$-equation by $R_m$, subtract, and integrate, we are in fact **proving self-adjointness in disguise**.

The integration by parts eliminates derivatives and leaves only a boundary term, which vanishes because of the imposed boundary conditions (Dirichlet–Dirichlet, in this case).

So the textbook derivation of orthogonality is essentially an explicit check of the abstract property “$\mathcal{L}$ is self-adjoint under this inner product.”

---

**SUMMARY**

* The **deep reason**: orthogonality follows from the Sturm–Liouville operator being self-adjoint under the weighted inner product $\int r f g ,dr$.
* The **calculation trick**: multiplying two eigenvalue equations, subtracting, and integrating is a constructive way to demonstrate this property explicitly.
* The **physical meaning**: the measure $r,dr$ corresponds to cylindrical area elements, so orthogonality is not just a formal result — it ensures that eigenmodes are independent when integrated over the disk area.
* Self-adjointness means nothing more (and nothing less) than:
    * You have a linear operator $\mathcal{L}$ on a function space.
    * You equip that space with an inner product $\langle f,g\rangle$.
        * The **inner product** acts like a **metric** on the function space: it defines lengths ($|f|^2=\langle f,f\rangle$) and angles (through orthogonality).
    * If for **all** admissible $f,g$ one has
    $$
    \langle f, \mathcal{L}[g]\rangle = \langle \mathcal{L}[f], g\rangle,
    $$
    then $\mathcal{L}$ is called **self-adjoint** (or Hermitian, in finite-dimensional linear algebra).
    * Why it matters:
        * Self-adjoint operators have **real eigenvalues**.
        * Their eigenfunctions are **orthogonal** with respect to the chosen inner product.
        * So saying that an operator $\mathcal{L}$ is **self-adjoint** means it is “symmetric” with respect to that metric.
:::

**APPLICATION TO THE ANNULAR DISK (DIRICHLET-DIRICHLET)**

For the annulus, the eigenfunctions are

$$
F_n(r) = J_0(\lambda_n r) + b_n Y_0(\lambda_n r),
$$

with $\lambda_n$ satisfying the characteristic equation

$$
J_0(\lambda_n R),Y_0(\lambda_n R_i) - J_0(\lambda_n R_i),Y_0(\lambda_n R) = 0.
$$

The orthogonality condition becomes

$$
\int_{R_i}^{R} r{}F_m(r){}F_n(r){}dr = 0, \qquad m\neq n.
$$

* Here, $F_n(r)$ includes both $J_0$ and $Y_0$, because the domain does not include the singularity at $r=0$.
* The boundary conditions $F_n(R_i)=0$, $F_n(R)=0$ ensure the boundary term in the derivation vanishes, just as in the solid-disk case.
* The result: the eigenfunctions ${F_n(r)}$ form an orthogonal set with respect to the weight $r$ over the annulus.

---

**NORMALIZATION CONSTANT FOR THE ANNULUS**

For each mode $n$, define the norm

$$
N_n = \int_{R_i}^{R} r\left[F_n(r)\right]^2{}dr.
$$

Then the orthogonality and normalization conditions are

$$
\int_{R_i}^{R} r{}F_m(r){}F_n(r){}dr = N_n\delta_{mn}.
$$

Unlike the solid-disk case, where $N_n$ has a simple closed form in terms of $J_1(j_{0,n})$, here the integral involves both $J_0$ and $Y_0$ and is typically evaluated numerically. Nevertheless, the Sturm–Liouville structure guarantees $N_n$ is finite and positive.

---

**HOW IT IS USED IN PRACTICE**

Suppose at $x=0$ the actuator disk induces an axial velocity distribution $v_x(r,0^+)=g(r)$ (from continuity, swirl, and pressure drop). To expand this onto eigenfunctions:

$$
g(r) = \sum_{n=1}^\infty A_n{}F_n(r).
$$

The orthogonality relation with weight $r$ allows the projection:

$$
A_n = \frac{1}{N_n}\int_{R_i}^{R} r{}g(r){}F_n(r){}dr.
$$

This is the annular generalization of the projection formula you had for the solid disk. The weight $r$ is essential: it comes directly from the divergence operator in cylindrical coordinates.

---

**PHYSICAL INTERPRETATION**

* Orthogonality means the radial eigenmodes represent **independent “channels”** of energy redistribution in the flow: they do not overlap under the natural inner product of the problem.
* The weight $r$ reflects the geometry: in cylindrical coordinates, area elements are $r,dr,d\theta$, so orthogonality is naturally defined with this measure.
* For the annular disk, the presence of both $J_0$ and $Y_0$ allows modes that satisfy two boundaries, and orthogonality ensures that these mixed-function modes still form a complete and independent set for expansions in the annular domain.

::::

::::{admonition} On the roles of $b_n$ and $F_{n}(r)$
:class: note

**THE ROLE OF $b_n$**

The general radial solution is

$$
F_n(r) = J_0(\lambda_n r) + b_n{}Y_0(\lambda_n r).
$$

* The constant $b_n$ tells you how much of the $Y_0$ component is mixed into the eigenfunction.
* Its value is fixed by the **boundary conditions**.

For example, in the annular disk with Dirichlet–Dirichlet conditions ($F_n(R_i)=0$, $F_n(R)=0$):

$$
J_0(\lambda_n R) + b_n{}Y_0(\lambda_n R) = 0,
\qquad
J_0(\lambda_n R_i) + b_n{}Y_0(\lambda_n R_i) = 0.
$$

Eliminating $b_n$ gives the **characteristic equation** for $\lambda_n$:

$$
J_0(\lambda_n R){}Y_0(\lambda_n R_i) - J_0(\lambda_n R_i){}Y_0(\lambda_n R)=0.
$$

Once $\lambda_n$ is found, $b_n$ follows from either boundary condition, e.g.

$$
b_n = -\frac{J_0(\lambda_n R)}{Y_0(\lambda_n R)}.
$$

So $b_n$ is an **intrinsic property of each mode shape**, determined by geometry and eigenvalue $\lambda_n$.

---

**THE ORTHOGONALITY CONDITION**

The **orthogonality** of eigenfunctions is a structural property of the Sturm–Liouville problem:

$$
\int_{R_i}^R r{}F_m(r){}F_n(r){}dr = 0 \quad (m\neq n).
$$

This relation does **not** determine $b_n$ or $\lambda_n$. Instead:

* It guarantees that the set ${F_n}$ forms a basis (like sines and cosines in Fourier series).
* It allows projection of an arbitrary inlet velocity profile $g(r)$ onto the eigenmodes, with coefficients
  $$
  A_n = \frac{1}{N_n}\int_{R_i}^{R} r{}g(r){}F_n(r){}dr,
  \qquad
  N_n=\int_{R_i}^R r{}[F_n(r)]^{2}{}dr.
  $$

So orthogonality is what lets us **expand arbitrary fields in this eigenbasis**, while $b_n$ is what defines each eigenfunction.

---

**RELATIONSHIP BETWEEN THEM**

* $b_n$ **belongs to the definition of each eigenfunction**. Without fixing $b_n$, you don’t know the actual shape of $F_n(r)$.
* Orthogonality is a **property of the entire set of eigenfunctions ${F_n}$**, once each $b_n$ has been determined.

In other words:

* **$b_n$ comes from boundary conditions.**
* **Orthogonality comes from the Sturm–Liouville structure.**

::::

---

**PHYSICAL INTERPRETATION OF THE CHARACTERISTIC EQUATION**

* The determinant condition encodes the requirement that both boundary conditions (hub and shroud walls) are simultaneously satisfied.
* Each root $\lambda_n$ corresponds to a radial standing-wave pattern between $R_i$ and $R$.
* Larger $\lambda_n$ correspond to finer radial structure and stronger exponential decay in $x$, meaning these modes dissipate faster downstream.
* The lowest $\lambda_1$ represents the most persistent, large-scale distortion of axial velocity induced by the disk.

Thus, the separation of variables not only gives the general solution form, but also reveals how the **discrete spectrum of $\lambda_n$** arises from the geometry and boundary conditions — the hallmark of a Sturm–Liouville problem.

---

**SUMMARY**

The PDE is essentially the **Laplace equation in a semi-infinite cylindrical geometry**, describing how axial-velocity disturbances generated at the actuator disk decay downstream. Its solution naturally decomposes into **Bessel eigenmodes radially** and **exponentially decaying modes axially**. The physics is that disturbances with smaller radial scale decay more quickly, while large-scale (small $\lambda$) structures persist further downstream. The coefficients $A_n$ are linked to the disk’s loading through swirl and pressure-jump conditions, making the mathematical expansion directly tied to the aerodynamics of the actuator disk.

# ON THE IMPOSITION OF THE BOUNDARY CONDITIONS

**RECAP**

* We are solving the axisymmetric Laplace equation for the axial disturbance

    $$
    \frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial v_x}{\partial r}\right)
    +\frac{\partial^2 v_x}{\partial x^2}=0,
    $$
    
    with separation of variables $v_x(r,x)=R(r)X(x)$.

* This gives:

    * Radial ODE:
      $$
      \frac{1}{r}\frac{d}{dr}\left(r\frac{dR}{dr}\right)+\lambda^2 R=0,
      $$
    
    * Axial ODE:
      $$
      \frac{d^2X}{dx^2}-\lambda^2 X=0.
      $$

* Solutions:

    * Radial:
      $$
      R(r)=J_0(\lambda r)+b,Y_0(\lambda r),
      $$
    
    * Axial:
      $$
      X(x)=e^{-\lambda |x|}.
      $$

* So the general mode is

    $$
    v_{x, n}(r,x)=A_n{}e^{-\lambda_n|x|}F_n(r),
    \qquad
    F_n(r)=J_0(\lambda_n r)+b_n{}Y_0(\lambda_n r).
    $$

---

**EIGENVALUE CONDITION: WALL IMPERMEABILITY**

* **Physical requirement:**

    The cylindrical walls at $r=R_i$ (hub) and $r=R$ (shroud) are solid → radial velocity vanishes:

    $$
    v_r(R_i,x)=0,\qquad v_r(R,x)=0.
    $$

* **Relation to eigenfunction:**

    With a potential representation $\phi(r,x)=R(r)X(x)$,
    
    $$
    v_r(r,x)=\frac{\partial \phi}{\partial r} = R'(r)X(x).
    $$
    
    Thus impermeability at the walls requires
    
    $$
    R'(R_i)=0,\qquad R'(R)=0.
    $$
    
    Equivalently, for each mode eigenfunction $F_n(r)$,
    
    $$
    F_n'(R_i)=0,\qquad F_n'(R)=0.
    $$

::::{admonition} On the Neumann-like condition for $F_n$
:class: note

**EXPRESSING $v_r$ IN THE SEPARATED REPRESENTATION IN THE SEPARATED REPRESENTATION**

Work with the (locally) irrotational meridional flow (the same assumption used to obtain the Laplace operator). Introduce a potential $\phi(r,x)$ so that

$$
v_r=\frac{\partial \phi}{\partial r},\qquad
v_x=\frac{\partial \phi}{\partial x}.
$$

Let the separated (single–mode) potential be

$$
\phi(r,x)=\Phi(r){}H(x),\qquad
\text{with}\quad \Phi\text{, }H \ \text{to be determined}.
$$

Then

$$
v_r(r,x)=\Phi'(r){}H(x),\qquad
v_x(r,x)=\Phi(r){}H'(x).
$$

Imposing impermeability at the two walls **for all $x$** gives

$$
\Phi'(R_i)=0,\qquad \Phi'(R)=0.
$$

Thus the **radial factor in the potential** must satisfy **Neumann** conditions at both walls.

---

**THE RADIAL FACTOR FOR $v_x$ IS THE SAME $\Phi(r)$**

The axial disturbance used in the expansion is (up to an overall $\pm\lambda$ factor)

$$
v_x(r,x)=\Phi(r){}H'(x).
$$

Therefore the **radial eigenfunction for $v_x$** is simply

$$
F(r)=\Phi(r).
$$

Because $\Phi'(R_i)=\Phi'(R)=0$, the **equivalent boundary condition for the radial eigenfunction of $v_x$** is

$$
F'(R_i)=0,\qquad F'(R)=0.
$$

---

**CONSISTENCY WITH THE BESSEL FORM OF $F_n$**

From separation of variables applied to Laplace’s equation, the radial eigenfunctions satisfy

$$
\frac{1}{r}\frac{d}{dr}\left(r\frac{dF}{dr}\right)+\lambda^2 F=0
\quad\Longrightarrow\quad
F(r)=J_0(\lambda r)+b{}Y_0(\lambda r).
$$

The Neumann wall conditions derived above become

$$
F'(R_i)= -\lambda\big[J_1(\lambda R_i)+b{}Y_1(\lambda R_i)\big]=0,
\quad
F'(R)=   -\lambda\big[J_1(\lambda R)+b{}Y_1(\lambda R)\big]=0.
$$

For a nontrivial eigenfunction, these two equations give the **characteristic equation** for $\lambda$ in the Neumann–Neumann annulus,

$$
J_1(\lambda R){}Y_1(\lambda R_i)-J_1(\lambda R_i){}Y_1(\lambda R)=0,
$$

and, once a root $\lambda=\lambda_n$ is found, the corresponding mixing coefficient is

$$
b_n=-\frac{J_1(\lambda_n R)}{Y_1(\lambda_n R)}
=-\frac{J_1(\lambda_n R_i)}{Y_1(\lambda_n R_i)}.
$$

---

**WHY THIS IS THE *AXIAL-VELOCITY* BOUNDARY CONDITION (NOT DIRICHLET)**

* The physical wall condition is $v_r=0$ (impermeability).
* In the separated potential form, $v_r\propto \Phi'(r)$, so the wall condition **translates to a Neumann condition on the radial factor**.
* Because $v_x\propto \Phi(r)$, **the same Neumann condition applies to the radial eigenfunction $F(r)$ of $v_x$**, i.e. $F'(R_i)=F'(R)=0$.

::::

* **Characteristic equation:**

    Since
    
    $$
    F_n'(r)=-\lambda_n \big[J_1(\lambda_n r)+b_n Y_1(\lambda_n r)\big],
    $$
    
    the boundary conditions imply
    
    $$
    J_1(\lambda_n R_i)+b_n Y_1(\lambda_n R_i)=0,\
    J_1(\lambda_n R)+b_n Y_1(\lambda_n R)=0.
    $$
    
    For a nontrivial $b_n$, the determinant condition is
    
    $$
    J_1(\lambda_n R),Y_1(\lambda_n R_i)-J_1(\lambda_n R_i),Y_1(\lambda_n R)=0.
    $$
    
    This transcendental equation gives the admissible eigenvalues $\lambda_n$. Then $b_n$ is fixed as
    
    :::{math}
    :label: eq_neumann_cond
    b_n=-\frac{J_1(\lambda_n R)}{Y_1(\lambda_n R)}
    =-\frac{J_1(\lambda_n R_i)}{Y_1(\lambda_n R_i)}.
    :::

    Mind that this condition is different from the one stated in {eq}`eq_dirichlet_cond`, as it is no more a Dirichlet condition, but a Neumann one.

So the wall impermeability condition leads to the determination of the $(\lambda_n,b_n)$ pairs.

---

**CONTINUITY OF AXIAL VELOCITY AT THE DISK PLANE**

* **Physical requirement:**

    The *total* axial velocity is continuous across the infinitely thin actuator disk at $x=0$.
    * Upstream: background velocity $U$ plus disturbance $v_x(r,-0)$.
    * Downstream: background velocity $V_{+\infty x}(r)$ plus disturbance $v_x(r,+0)$.
    Hence, the continuity condition reads:
    $$
    U+v_x(r,-0)=V_{+\infty x}(r)+v_x(r,+0).
    $$

* **Oddness of the disturbance:**

  From the exponential decay form $e^{-\lambda |x|}$, the disturbance is even in $|x|$; to keep total velocity continuous, we impose oddness across the disk:

    $$
    v_x(r,-0)=-v_x(r,+0).
    $$

* **Derivation:**

    Insert oddness into continuity condition:

    $$
    U+v_x(r,-0)=V_{+\infty x}(r)-v_x(r,-0),
    $$
    
    so
    
    $$
    2v_x(r,-0)=V_{+\infty x}(r)-U,
    $$
    
    that is
    
    $$
    v_x(r,-0)=\tfrac{1}{2}\big[V_{+\infty x}(r)-U\big],\qquad
    v_x(r,+0)=-\tfrac{1}{2}\big[V_{+\infty x}(r)-U\big].
    $$

---

**EXPANSION ONTO EIGENFUNCTIONS (FIXING $A_n$)**

* **Imposing the functional shape**

    At $x=0$, each eigenmode reduces to its radial part:
    $$
    v_x(r,\pm 0)=\pm \sum_{n=1}^\infty A_n F_n(r).
    $$

    Using the convention

    $$
    v_x(r,-0)=-\sum_{n=1}^\infty A_n F_n(r),
    $$
    
    the continuity condition gives
    
    $$
    -\sum_{n=1}^\infty A_n F_n(r) = \tfrac{1}{2}\big[V_{+\infty x}(r)-U\big].
    $$

* **Projecting by using orthogonality**


    Since ${F_n(r)}$ form an orthogonal set with weight $r$:

    $$
    \int_{R_i}^R r{}F_m(r)F_n(r){}dr = N_n \delta_{mn},
    $$

    we can solve for each coefficient:
    
    $$
    A_n = -\frac{1}{N_n}\int_{R_i}^R r{}\frac{1}{2}\big[V_{+\infty x}(r)-U\big]F_n(r){}dr.
    $$

---

**WHY THE CONDITION FOR THE RADIAL GRADIENT OF THE AXIAL DISTURBANCE IS NEEDED**

After separation of variables the axial disturbance is

$$
v_x(r,x)=\sum_{n=1}^\infty A_n,e^{-\lambda_n|x|},F_n(r),
$$

with the radial eigenpairs $(\lambda_n,b_n)$ fixed by **wall impermeability** (Neumann at $r=R_i,R$: $F_n'(R_i)=F_n'(R)=0$).

The **continuity of total axial velocity** at the disk plane gives the *value* of the disturbance on the two sides,

$$
v_x(r,-0)=-v_x(r,+0)=\tfrac12\big[V_{+\infty x}(r)-U\big].
$$

These two ingredients alone do **not** yet determine the amplitudes ${A_n}$, because the downstream profile $V_{+\infty x}(r)$ (or, equivalently, the right-hand side above) is **not known a priori**.

A third relation is required: the **radial component of Euler’s equation** at $x=0^+$, which links the *radial derivative* of $v_x$ to the swirl field and the stagnation-pressure drop. That is Eq. (17).

---

**DERIVING THE BOUNDARY CONDITION EQUATION FOR THE RADIAL GRADIENT OF THE AXIAL DISTURBANCE**

We work at the **exit of the disk**, $x=0^+$, in steady, inviscid, axisymmetric flow. Cylindrical components are $(v_r,v_\theta,v_x)$; $p$ is static pressure; $\rho$ is density.

* **Radial Euler (steady, axisymmetric)**

    The exact radial momentum balance is
    
    :::{math}
    :label: eq_radial_momentum
    v_r\frac{\partial v_r}{\partial r}
    +v_x\frac{\partial v_r}{\partial x}
    -\frac{v_\theta^2}{r}
    = -\frac{1}{\rho}\frac{\partial p}{\partial r}.
    :::

* **Near-disk linearization and irrotationality**

    Near $x=0^+$, take $v_r$ as small (no radial mass flux through the sheet) and neglect the quadratic term $v_r\partial_r v_r$.
    
    Assume the meridional flow is locally irrotational, so mixed partials commute:
    
    $$
    \frac{\partial v_r}{\partial x}=\frac{\partial v_x}{\partial r}.
    $$
    
    With these, {eq}`eq_radial_momentum` becomes
    
    :::{math}
    :label: eq_radial_momentum_with_irrotational_condition
    v_x(r,0)\frac{\partial v_x}{\partial r}(r,0^+)-\frac{v_\theta^2(r,0^+)}{r}
    =-\frac{1}{\rho}\frac{\partial p}{\partial r}(r,0^+).
    :::
    
    Here $v_x(r,0)$ is the **mean axial speed at the plane** (the average of upstream $U$ and downstream $V_{+\infty x}(r)$.

::::{admonition} Irrotationality, shear, and rotation
:class: note

When studying fluid motion, it is useful to think about what happens to a **tiny fluid element** (say, a square in the $(x,y)$ plane). The local velocity gradient $\nabla \mathbf{v}$ determines how this element deforms and moves.

**SHEAR VS ROTATION**

The velocity gradient can be decomposed into two parts:

* **Symmetric part (strain rate)**:
  $$
  \mathbf{E} = \tfrac12\big(\nabla \mathbf{v} + (\nabla \mathbf{v})^\top\big),
  $$
  which describes stretching, compression, and *shear deformation*.

* **Antisymmetric part (rotation rate)**:
  $$
  \mathbf{\Omega} = \tfrac12\big(\nabla \mathbf{v} - (\nabla \mathbf{v})^\top\big),
  $$
  which corresponds to rigid-body rotation, and is linked directly to the vorticity.

**THE IRROTATIONALITY CONDITION**

In the meridional $(r,x)$ plane, the irrotationality condition is
$$
\frac{\partial v_r}{\partial x} = \frac{\partial v_x}{\partial r}.
$$

This ensures that the **antisymmetric part vanishes**: the element does not spin. Circulation around a tiny rectangular loop cancels out, and the vorticity component normal to the plane is zero.

But this does *not* mean that there is no shear. The symmetric part $\mathbf{E}$ may still be nonzero: the element can be stretched or sheared into a parallelogram. The key point is that **shear without vorticity produces deformation without rotation**.

**MENTAL PICTURE**

* If the antisymmetric part is present ($\Omega\neq 0$), the element spins like a little rigid body.
* If the symmetric part is present ($E\neq 0$), the element deforms — sides change length, angles distort.
* If only $\Omega$ is present → pure rigid-body rotation.
* If only $E$ is present → pure strain, with no spin.
* If both are present → the element both rotates and deforms.
* If neither is present → the element just translates.

So: **Irrotational ≠ shear-free**. It simply means that shear contributions balance so that the element doesn’t spin, even if it still deforms.

---

**TOY CASE**

* **Irrotational shear (shear without rotation)**

    * Take a shear field where the $x$-velocity varies linearly with $y$:
        $$
        v_x(x,y) = sy, \qquad v_y(x,y) = 0,
        $$
        with shear rate $s(t)$ oscillating in time.
    
    * Computing the velocity matrix yields:

        $$
        \nabla \mathbf{v} =
        \begin{bmatrix}
        \partial v_x/\partial x & \partial v_x/\partial y \\
        \partial v_y/\partial x & \partial v_y/\partial y
        \end{bmatrix} =
        \begin{bmatrix}
        0 & s \\
        0 & 0
        \end{bmatrix}.
        $$

    * The symmetric part (strain rate) is:

      $$
      \mathbf{E} = \tfrac12(\nabla \mathbf{v} + \nabla \mathbf{v}^\top) =
      \frac12\begin{bmatrix}
      0 & s \\
      0 & 0
      \end{bmatrix} + 
      \frac12\begin{bmatrix}
      0 & 0 \\
      s & 0
      \end{bmatrix} =
      \frac12\begin{bmatrix}
      0 & s \\
      s & 0
      \end{bmatrix}.
      $$

    * The antisymmetric part (rotation rate) is:

      $$
      \mathbf{\Omega} = \tfrac12(\nabla \mathbf{v} - \nabla \mathbf{v}^\top) =
      \frac12\begin{bmatrix}
      0 & s \\
      0 & 0
      \end{bmatrix} - 
      \frac12\begin{bmatrix}
      0 & 0 \\
      s & 0
      \end{bmatrix} =      
      \frac12\begin{bmatrix}
      0 & s \\
      -s & 0
      \end{bmatrix}.
      $$

    * For *purely irrotational shear*, we want $\Omega=0$. That requires the cross-derivative condition
      $$
      \frac{\partial v_x}{\partial y} = \frac{\partial v_y}{\partial x},
      $$
      which here means $s=0$ unless accompanied by a compensating $v_y$.


* **Pure rigid-body rotation**

    * Take a uniform rotation about the origin with angular velocity $\omega$:

        $$
        v_x(x,y) = -\omega y, \qquad v_y(x,y) = \omega x.
        $$

    * The velocity gradient is:

        $$
        \nabla \mathbf{v} =
        \begin{bmatrix}
        0 & -\omega \\
        \omega & 0
        \end{bmatrix}.
        $$

    * The symmetric part is:

      $$
      \mathbf{E} = \tfrac12
      \begin{bmatrix}
      0 & -\omega \\
      \omega & 0
      \end{bmatrix} +
      \tfrac12
      \begin{bmatrix}
      0 & \omega \\
      -\omega & 0
      \end{bmatrix}
      = \begin{bmatrix} 0 & 0 \\
      0 & 0
      \end{bmatrix}.
      $$

    * The antisymmetric part is:

      $$
      \mathbf{\Omega} = \tfrac12
      \begin{bmatrix}
      0 & -\omega \\
      \omega & 0
      \end{bmatrix} -
      \tfrac12
      \begin{bmatrix}
      0 & \omega \\
      -\omega & 0
      \end{bmatrix} =
      \begin{bmatrix}
      0 & -\omega \\
      \omega & 0
      \end{bmatrix}.
      $$

    * So here **only rotation, no shear**.

* **Rotation + shear (general case)**

    * Take a superposition: rigid rotation plus a shear term in $x$ proportional to $y$:

      $$
      v_x(x,y) = -\omega y + s y, \qquad v_y(x,y) = \omega x.
      $$

    * The velocity gradient is:

      $$
      \nabla \mathbf{v} =
      \begin{bmatrix}
      0 & -\omega + s \\
      \omega & 0
      \end{bmatrix}.
      $$

    * The symmetric part is:

      $$
      \mathbf{E} =
      \tfrac12\begin{bmatrix}
      0 & -\omega+s \\
      \omega & 0
      \end{bmatrix} +
      \tfrac12\begin{bmatrix}
      0 & \omega \\
      -\omega+s & 0
      \end{bmatrix} =
      \frac12
      \begin{bmatrix}
      0 & s \\
      s & 0
      \end{bmatrix}.
      $$

    * The antisymmetric part:

      $$
      \mathbf{\Omega} =
      \tfrac12\begin{bmatrix}
      0 & -\omega+s \\
      \omega & 0
      \end{bmatrix} -
      \tfrac12\begin{bmatrix}
      0 & \omega \\
      -\omega+s & 0
      \end{bmatrix} =
      \begin{bmatrix}
      0 & -\omega + \frac{s}{2} \\
      \omega - \frac{s}{2} & 0
      \end{bmatrix}.
      $$
    
    * So the shear part contributes to $\mathbf{E}$, the rotation part to $\mathbf{\Omega}$.

* **Introduce stagnation pressure**

    Define the downstream stagnation pressure (including swirl kinetic energy):
    
    $$
    p_0^+(r)=p(r,0^+)+\tfrac12\rho\big(v_x^2(r,0^+)+v_\theta^2(r,0^+)\big).
    $$
    
    Differentiate:
    
    $$
    \frac{1}{\rho}\frac{d p_0^+}{dr}
    =\frac{1}{\rho}\frac{d p}{dr}
    +v_x\frac{d v_x}{dr}
    +v_\theta\frac{d v_\theta}{dr}.
    $$
    
    Use {eq}`eq_radial_momentum_with_irrotational_condition` to substitute $(1/\rho),dp/dr = -,v_x,dv_x/dr + v_\theta^2/r$, giving
    
    :::{math}
    :label: eq_radial_equilibrium_relation_swirling_flow
    \frac{1}{\rho}\frac{d p_0^+}{dr}
    =\frac{v_\theta^2}{r}+v_\theta\frac{d v_\theta}{dr}
    =\frac{v_\theta}{r}\frac{d}{dr}\big(r,v_\theta\big),
    \quad x=0^+.
    :::
    
    This is the **radial equilibrium relation** for a swirling flow.

* **Relating to stagnation-pressure drop**

    Let $\Delta p_0(r)$ be the stagnation-pressure drop across the disk:
    
    $$
    \Delta p_0(r)=p_0^-(r)-p_0^+(r).
    $$
    
    Then
    
    :::{math}
    :label: eq_pressure_differential_up_down
    \frac{1}{\rho}\frac{d p_0^+}{dr}
    =\frac{1}{\rho}\frac{d p_0^-}{dr}-\frac{1}{\rho}\frac{d\Delta p_0}{dr}.
    :::
    
    Assume upstream stagnation pressure is radially uniform, so $dp_0^-/dr=0$. From {eq}`eq_radial_equilibrium_relation_swirling_flow`–{eq}`eq_pressure_differential_up_down`:
    
    $$
    \frac{v_\theta}{r}\frac{d}{dr}\big(r{}v_\theta\big)
    = -\frac{1}{\rho}\frac{d\Delta p_0}{dr}.
    $$

* **Solve for $dv_x/dr$**

    From {eq}`eq_radial_momentum_with_irrotational_condition`:
    
    :::{math}
    :label: eq_radial_equilibrium_simplified
    v_x(r,0)\frac{d v_x}{dr}(r,0^+)
    =\frac{v_\theta^2}{r}-\frac{1}{\rho}\frac{d p}{dr}(r,0^+).
    :::
    
    But
    
    $$
    \frac{1}{\rho}\frac{d p}{dr}
    =\frac{1}{\rho}\frac{d p_0^+}{dr} - 
    v_x\frac{d v_x}{dr} - 
    v_\theta\frac{d v_\theta}{dr}.
    $$
    
    Insert this into {eq}`eq_radial_equilibrium_simplified`, use {eq}`eq_radial_equilibrium_relation_swirling_flow`, and rearrange. The result is
    
    $$
    \frac{d v_x}{dr}(r,0^+)
    =\frac{1}{2{}v_x(r,0)}
    \left[
    \frac{v_\theta(r,0^+)}{r}\frac{d}{dr}\big(r{}v_\theta(r,0^+)\big)
    -\frac{1}{\rho}\frac{d\Delta p_0}{dr}(r)
    \right].
    $$

* **Mean axial velocity at the disk plane**

    From continuity of **total** axial velocity at $x=0$:
    
    $$
    v_x(r,-0)=-v_x(r,+0)=\tfrac12\big[V_{+\infty x}(r)-U\big],
    $$
    
    so the mean axial velocity at the disk plane is
    
    $$
    v_x(r,0)=\tfrac12\big(U+V_{+\infty x}(r)\big).
    $$

---

**UNIFYING NOTATION**

* Upstream total axial velocity (far): $U$.

* Downstream total axial velocity (far): $V_{+\infty x}(r)$.

* Axial **disturbance** (perturbation) near the disk: $v_x(r,x)$ with
  $$
  v_x(r,-0)=-v_x(r,+0)=\tfrac12\big[V_{+\infty x}(r)-U\big].
  $$

* Mean **total** axial velocity at the plane (the figure’s $V_x(r,0)$):
  $$
  V_x(r,0)=\tfrac12\big[U+V_{+\infty x}(r)\big].
  $$

* Tangential velocity at $x=0^+$. Two options:

  * If upstream swirl is negligible (common in actuator-disk use): total swirl equals the **perturbation**,
    $$
    V_\theta(r,0)\equiv V_\theta(r,0^+)=v_\theta(r,0^+).
    $$
  * If an upstream base swirl $V_{\theta0}^-(r)$ exists, then the **total** swirl at the plane is
    $$
    V_\theta(r,0)=V_{\theta0}^-(r)+v_\theta(r,0^+).
    $$
* $\Delta p_0(r)=p_0^-(r)-p_0^+(r)$: stagnation-pressure drop across the disk.

* **Most common case (no upstream swirl: $V_{\theta{}0}^-=0$)**

    Then $V_\theta(r,0)=v_\theta(r,0^+)$, and
    $$
    \boxed{%
    \frac{\partial v_x}{\partial r}(r,0^+)
    =\frac{1}{U+V_{+\infty x}(r)}
    \left[
    \frac{v_\theta(r,0^+)}{r}
    \frac{d}{dr}\big(rv_\theta(r,0^+)\big)
    -\frac{1}{\rho}\frac{d}{dr}\Delta p_0(r)\right]
    }
    $$
    since $2V_x(r,0)=U+V_{+\infty x}(r)$.

* **General case (allow upstream base swirl $V_{\theta{}0}^-(r)$)**

    Use the **total** swirl at the plane,
    $$
    V_\theta(r,0)=V_{\theta0}^-(r)+v_\theta(r,0^+),
    $$
    and the same mean axial velocity $V_x(r,0)=\tfrac12[U+V_{+\infty x}(r)]$. Then
    $$
    \boxed{%
    \frac{\partial v_x}{\partial r}(r,0^+)
    =\frac{1}{U+V_{+\infty x}(r)}
    \left[
    \frac{V_\theta(r,0)}{r}
    \frac{d}{dr}\big(r{}V_\theta(r,0)\big)
    -\frac{1}{\rho}\frac{d}{dr}\Delta p_0(r)\right]
    }
    $$
    with $V_\theta(r,0)$ defined as above. Expanding $V_\theta$ gives explicit cross-terms between the upstream swirl and the downstream perturbation if you need them.

* **Obtaining the final form**

    * Replace $U+V_{+\infty x}(r)$ by $2V_x(r,0)$.
    * Replace $v_\theta(r,0^+)$ (or $V_\theta(r,0)$ in the general case).

        $$
        \frac{d}{dr}v_x(r,0^+)
        =\frac{1}{2V_x(r,0)}
        \frac{V_\theta(r,0)}{r}
        \frac{d}{dr}\big[r{}V_\theta(r,0)
        -\frac{1}{\rho}\frac{d}{dr}\Delta p_0(r)\big].
        $$

---

**SUMMARY**

* **Eigenvalue pairs $(\lambda_n,b_n)$** are determined solely by the **impermeable wall condition** $v_r=0$, which is equivalent to $F_n'(R_i)=F_n'(R)=0$.
    * **Eigenvalue condition:** defines the shape of admissible modes.
* **Mode amplitudes $A_n$** are determined by the **continuity of axial velocity across the disk plane**, which matches the jump between upstream $U$ and downstream $V_{+\infty x}(r)$ via Eq. (16a–b).
    * **Continuity condition:** fixes the amplitudes of those modes for a given physical loading.
* Together, they yield the full modal expansion:
  $$
  v_x(r,x)=\sum_{n=1}^\infty A_n{}e^{-\lambda_n|x|}{}F_n(r).
  $$
* **Radial Euler at $x=0^+$** $;\Rightarrow$ closure: determines $dv_x/dr$ at the disk in terms of swirl $v_\theta$ and stagnation-pressure drop $\Delta p_0$. This final condition lets us project onto eigenfunctions and compute the amplitudes $A_n$.

(sec_on_the_iteration_loop)=
# ON THE ITERATION LOOP

**GEOMETRY → EIGENVALUES AND EIGENFUNCTIONS**

* Once the hub radius $R_i$ and shroud radius $R$ are fixed, the **radial eigenproblem** is fully determined.
* Solve the Neumann boundary condition $F_n'(R_i)=F_n'(R)=0$ once to obtain the eigenvalues $\lambda_n$ and the mixing ratios $b_n$.
* This gives you the **radial eigenfunctions** $F_n(r)$ that will never change in the iteration.

---

**INITIALIZE WITH ANA XIAL VELOCITY GUESS**

* Assume a **radial distribution** for the axial velocity at the disk plane, $V_x(r,0)$.
* This is typically written as upstream uniform velocity plus disturbance:
  $$
  V_x(r,0)=U+v_x(r,-0)=\tfrac12\big[U+V_{+\infty x}(r)\big].
  $$

---

**CASCADE DATA: SWIRL AND PRESSURE DROP**

* From a **two-dimensional cascade model**, compute:

  * The **stagnation-pressure drop** $\Delta p_0(r)$ imposed by the blade row.
  * The **swirl velocity downstream** of the cascade, $V_\theta(r,0^+)$.

These quantities are treated as *known inputs* to the actuator-disk formulation.

---

**USE RADIAL EULER EQUATION TO COMPUTE $\partial_r v_x(r,0^+)$**

* Insert $V_x(r,0)$, $V_\theta(r,0^+)$, and $\Delta p_0(r)$:
  $$
  \frac{\partial v_x}{\partial r}(r,0^+)
  =\frac{1}{2V_x(r,0)}\left[\frac{V_\theta(r,0)}{r}\frac{d}{dr}(rV_\theta(r,0))
  -\frac{1}{\rho}\frac{d}{dr}\Delta p_0(r)\right].
  $$

This gives the **radial derivative of the axial disturbance** as a function of $r$.

---

**EXPAND IN EIGENFUNCTIONS AND PROJECT**

* At $x=0^+$ the expansion is
  $$
  \frac{\partial v_x}{\partial r}(r,0^+)=\sum_{n=1}^\infty A_n{}F_n'(r).
  $$
* Using **orthogonality of the eigenfunctions**, project onto $F_m'(r)$ with weight $r$:
  $$
  A_m=\frac{1}{\lambda_m^2N_m}\int_{R_i}^{R} r\left(\frac{\partial v_x}{\partial r}(r,0^+)\right)F_m'(r),dr.
  $$

This gives the **modal amplitudes** $A_n$.

---

**UPDATE PERTURBED AXIAL VELOCITY**

* With the new $A_n$, reconstruct the axial disturbance everywhere:
  $$
  v_x(r,x)=\sum_{n=1}^\infty A_n{}e^{-\lambda_n|x|}{}F_n(r).
  $$

* In particular, at the disk plane $x=0^+$ you get
  $$
  v_x(r,0^+)=\sum_{n=1}^\infty A_n{}F_n(r).
  $$

* Update the downstream axial velocity profile:
  $$
  V_{+\infty x}(r)=U-2{}v_x(r,0^+).
  $$

---

**ITERATE**

* With this updated $V_{+\infty x}(r)$, recompute $V_x(r,0)$, then return to Step 3 to re-evaluate the pressure drop and swirl from cascade correlations.
* Repeat until $V_x(r,0)$, $V_\theta(r,0^+)$, and $\Delta p_0(r)$ are all self-consistent with the eigenfunction expansion.

---

**SUMMARY**

* Geometry $\to$ eigenpairs $(\lambda_n,b_n)$ once and for all.
* Guess $V_x(r,0)$, get $\Delta p_0$ and $V_\theta$ from cascade.
* Impose orthogonality to compute $A_n$ and hence $v_x(r,x)$.
* Update the downstream axial velocity profile.
* Loop until convergence.

(sec_on_the_radial_equilibrium_condition)=
# ON THE RADIAL EQUILIBRIUM CONDITION

**WHAT THE “BESSEL-FUNCTION EXPANSION” IS DOING**

* We solve the **radial part of Laplace’s equation** (or equivalently, the radial equilibrium condition derived from linearized Euler + continuity).
* The eigenfunctions $F_n(r)$ (combinations of $J_0$ and $Y_0$) form a complete basis for disturbances in an annulus.
* By expanding the disturbance $v_x(r,x)$ in this basis, we enforce **regularity in the interior** and **boundary conditions at $r=\left(R_i, R\right)$**.

So the modal expansion is not arbitrary — it’s exactly the mathematical way of satisfying radial equilibrium given the disk geometry.

---

**THE BOUNDARY-CONDITION AT THE DISK**

The imposed equation comes directly from the **radial momentum equation across the disk**:
$$
\frac{\partial v_x}{\partial r}(r,0^+) =
\frac{1}{2V_x(r,0)} \left[ \frac{V_\theta(r,0)}{r}\frac{d}{dr}(rV_\theta(r,0))\frac{1}{\rho}\frac{d}{dr}\Delta p_0(r)\right].
$$

* The term with $V_\theta$ is the centrifugal effect of the swirl distribution.
* The term with $\Delta p_0$ is the radial gradient of the stagnation pressure drop.
* The denominator $2V_x(r,0)$ enforces the actuator-disk continuity constraint.

This condition **closes the problem**: it projects the physical swirl and pressure-drop effects onto the mathematical eigenbasis.

---

**WHY RADIAL EQUILIBRIUM?**

In a cascade or actuator-disk setting:

* Axial pressure balance is given by the disk momentum condition.
* Radial equilibrium is the condition that the **flow downstream does not accelerate radially unchecked**, but instead arranges itself so that the centrifugal forces of swirl and the axial pressure deficit balance consistently.

If you ignored this, the downstream solution would violate continuity or produce an inconsistent pressure field.

---

**SUMMARY**

So the Bessel-based solution:

* Provides a basis that **exactly satisfies the homogeneous radial equilibrium equation** (Laplace in $(r,x)$).
* Uses the disk boundary condition to **fix the modal amplitudes** $A_n$.
* Guarantees that the reconstructed $v_x(r,x)$ downstream is consistent with both **continuity** and **radial momentum balance**.
* Hence:
    * **Radial boundary conditions →** give you the eigenfunctions $F_n(r)$ (mathematical basis).
    * **Continuity across disk →** sets the global consistency between inflow, disk plane, and outflow (average velocity balance).
    * **Radial derivative →** provides the forcing that determines the exact modal coefficients $A_n$ when projected onto the eigenbasis.
    * Continuity across the disk by itself does not independently determine the modal coefficients but enforces that the modal expansion of the disturbance is consistent with the average velocity balance between upstream and downstream; the actual numerical values and relations among the coefficients are fixed only once you apply the radial derivative condition, which already contains the $2V_x(r,0)$ factor from continuity in its denominator, so the two conditions are coupled: continuity constrains the mean level and provides the denominator, while the radial equilibrium derivative supplies the projection that yields the definite set of $A_n$.


(sec_comparison_ciappi_formulation)=
# ON THE COMPARISON WITH CIAPPI'S FORMULATION

**COMMON ORIGIN**

Both models start from the **same governing equation**:

$$
\frac{\partial V_r}{\partial t} + V_r \frac{\partial V_r}{\partial r} + V_x \frac{\partial V_r}{\partial x} - \frac{V_\theta^2}{r} ;=; -\frac{1}{\rho} \frac{\partial p}{\partial r}.
$$

* This is the **radial momentum balance** in cylindrical coordinates.
* Combined with continuity, it relates the radial variations of axial velocity $V_x$, swirl $V_\theta$, and static/total pressure $p$.
* What differs is **how $p_{ta}(r)$ is treated** and **how $V_x(r)$ is represented**.

---

**PATH TO EACH MODEL**

* **Ciappi’s path**:

  1. Start from radial Euler eqn.
  2. Write in terms of axial and swirl velocity components.
  3. Assume **constant total pressure** → $dp_{ta}/dr=0$.
  4. Simplify to direct ODE linking $v_x^2$ and $(rv_\theta)^2$.
  5. Integrate radially and adjust with continuity.

* **Gato’s path**:

  1. Start from same Euler equation.
  2. Keep **non-zero pressure drop** across the disk, $\Delta p_0(r)$.
  3. Superimpose perturbation velocity field $v_x = U + v'_x$ satisfying Laplace’s equation.
  4. Represent solution as Bessel series → ensures hub/tip boundary conditions.
  5. Determine coefficients $A_n$ by projecting forcing term (which includes swirl and $\Delta p_0$).

---

**STARTING ASSUMPTIONS**

* **Ciappi**

  * Based on the radial momentum balance (Euler equation) in differential form.
  * Starts with an expression where the radial derivative of the total pressure $p_{ta}$ appears explicitly (eq. 4.126).
  * A key simplification is the assumption of **constant total pressure along the radius**. Under this hypothesis, the first term in eq. (4.126) disappears, leading to the simplified forms (4.127–4.129).
  * The result is a direct relation between the radial variation of $v_x^2$ and the variation of the swirl velocity (or equivalently, the square of the circulation).

* **Gato (Bessel, actuator–disk formulation)**

  * Starts from Laplace’s equation for an irrotational perturbation of the axial velocity in cylindrical coordinates.
  * The perturbation is expressed as an exponential decay in $x$ and radial dependence in terms of Bessel functions.
  * Crucial assumption: the **actuator disk produces a discontinuity in the axial velocity** ($v_x$), which is handled by enforcing continuity conditions across the disk and a pressure balance derived from Euler’s radial component.
  * Unlike Ciappi, it does not assume constant total pressure; instead, the pressure jump across the disk, $\Delta p_0$, is explicitly included as a physical input linked to cascade losses.

---

**MATHEMATICAL FORMULATION**

* **Ciappi**

  * Governing equation:
    $$
    \frac{d(v_{x2}^2)}{dr} = -\frac{1}{4\pi^2 r^2} \frac{d(2\pi r v_{\theta 2})^2}{dr}.
    $$
  * This is a first–order ODE in $r$. Its solution gives $v_x(r)$ once the swirl distribution $v_\theta(r)$ is known.
  * Integration introduces a constant of integration that is adjusted using continuity (matching inlet and outlet mass flow).
  * The formulation is **local and direct**, giving $v_x(r)$ from $v_\theta(r)$ with relatively simple numerical work.

* **Gato**

  * Uses a modal expansion for the axial perturbation:
    $$
    v_x(r,x) = \sum_{n=1}^\infty A_n e^{-\lambda_n x} F_n(r),
    $$
    with $F_n(r)$ combinations of $J_0$ and $Y_0$, and $\lambda_n$ radial eigenvalues fixed by boundary conditions at hub and tip.
  * The coefficients $A_n$ are determined by projecting the forcing term, derived from Euler’s radial equation, onto the Bessel basis:
    $$
    \frac{d}{dr} v_x(r,0^+) = \frac{1}{2V_x(r,0)} \frac{V_\theta(r,0)}{r} \frac{d}{dr}[r V_\theta(r,0)] - \frac{1}{\rho}\frac{d}{dr}\Delta p_0(r).
    $$
  * The solution requires solving an eigenvalue problem and a system of modal amplitudes rather than just integrating one ODE.

---

**PHYSICAL INTERPRETATION**

* **Ciappi**

  * Describes radial redistribution of axial velocity **without explicitly modeling a disk actuator**.
  * The assumption of constant total pressure implies that losses are uniform along the span, so the outlet $p_{ta}$ is flat.
  * This is a strong simplification: real turbines may show significant radial gradients in pressure total due to non-uniform loading.
  * It yields a quick way to enforce radial equilibrium, but may underestimate spanwise non-uniformities.

* **Gato**

  * Includes the actuator–disk explicitly: the axial velocity jump and pressure drop are part of the formulation.
  * The modal structure captures more complex spanwise variations, consistent with cylindrical geometry and boundary conditions at hub and tip.
  * The downstream velocity field is not given by a simple ODE but by a **balance between swirl, pressure drop, and continuity at the disk plane**.
  * Physically richer: predicts how disturbances decay downstream and redistribute radially.

---

**PRACTICAL CONSEQUENCES**

* **Ciappi**

  * Useful for fast implementation when one wants a simplified radial equilibrium condition.
  * Appropriate if one accepts the constant–$p_{ta}$ assumption and wants an efficient computation.
  * Best suited to parametric studies or when only a rough distribution is needed.

* **Gato**

  * More computationally demanding (eigenvalues, projection onto Bessel modes).
  * Better suited when actuator–disk effects must be captured: pressure jump, detailed redistribution of axial velocity, and downstream decay of perturbations.
  * Physically more general, capable of reproducing spanwise gradients and actuator effects observed in OWC Wells turbine rotors.

---

**COMPARISON TABLE**

| Aspect                     | Ciappi                                                                                          | Gato (Bessel, actuator–disk)                                                                        |
| -------------------------- | ----------------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------- |
| **Starting point**         | Radial component of Euler’s equation + assumption of constant $p_{ta}(r)$                       | Radial Euler equation + Laplace equation for perturbation of $v_x$                                  |
| **Main assumption**        | Constant total pressure along the blade span (simplifies $dp_{ta}/dr=0$)                        | Explicit actuator disk with finite pressure drop $\Delta p_0(r)$                                    |
| **Form of solution**       | First–order ODE: $; \frac{d(v_x^2)}{dr} = -\frac{1}{4\pi^2r^2} \frac{d(2\pi r v_\theta)^2}{dr}$ | Modal expansion: $; v_x(r,x) = \sum A_n e^{-\lambda_n x}F_n(r)$, with $F_n$ Bessel functions        |
| **Boundary conditions**    | Chosen by integration constant (adjusted to match inlet–outlet continuity)                      | Eigenvalues $\lambda_n$ from hub/tip boundary conditions, coefficients $A_n$ from projection        |
| **Treatment of pressure**  | Total pressure assumed uniform along $r$ → it drops out                                         | Pressure drop $\Delta p_0(r)$ explicitly enters Eq. (17)                                            |
| **Treatment of swirl**     | Swirl velocity distribution $v_\theta(r)$ prescribed → defines $dv_x^2/dr$                      | Swirl enters explicitly in forcing term of $G(r)$                                                   |
| **Complexity**             | Relatively simple to implement; integration along $r$                                           | More complex (modal decomposition, eigenvalues, orthogonality)                                      |
| **Physical meaning**       | Imposes a "flattened" $p_{ta}(r)$ → radial redistribution of $v_x$ is minimalistic              | Models actuator disk jump + radial redistribution; predicts downstream decay and spanwise gradients |
| **Continuity enforcement** | Done by adjusting integration constant so $\dot m_{in}=\dot m_{out}$                            | Implicitly enforced via Bessel expansion + projection of forcing onto modes                         |

---

**SIDE-BY-SIDE COMPARISON**

* **Route A (mean-only):** Ciappi’s radial-equilibrium ODE for the **mean** outlet field.
* **Route B (perturbation + eigenmodes):** Gato’s actuator-disk, Bessel-modal formulation for a **perturbation** superimposed on a uniform base flow.

    * Common starting point (cylindrical coordinates)

        **Radial momentum (steady, inviscid):**
        $$
        V_r\frac{\partial V_r}{\partial r} + V_x\frac{\partial V_r}{\partial x} - \frac{V_\theta^2}{r}
        = -\frac{1}{\rho}\frac{\partial p}{\partial r}.
        $$
        
        **Continuity (steady, incompressible, axisymmetric):**
        $$
        \frac{1}{r}\frac{\partial}{\partial r}(r V_r) + \frac{\partial V_x}{\partial x} = 0.
        $$
        
        We will work in the **near field of the disk**, evaluate things **on the disk plane**, and relate the **radial pressure gradient** to the **centrifugal term** and the **radial acceleration** (which we soon eliminate/approximate under each route’s assumptions).

    * **ROUTE A - CIAPPI (MEAN-FLOW ODE)**

        * Choice of variables and evaluation plane

            Work with **mean outlet** fields at the disk downstream plane:
            $$
            V_x(r) \equiv V_{x2}(r),\qquad V_\theta(r) \equiv V_{\theta2}(r),\qquad p_t(r)\equiv p_{t2}(r).
            $$

        * Key simplifying assumption

            Assume **constant total pressure** along the span at the downstream plane:
            $$
            \frac{d p_t}{dr} = 0.
            $$
            Since $p_t = p + \tfrac{1}{2}\rho(V_x^2 + V_\theta^2)$, this implies a specific relation between static-pressure gradient and the radial variation of the kinetic term; after standard manipulations (using continuity to eliminate $V_r$ at the plane and discarding axial-gradient terms at the plane), one arrives at the **radial equilibrium in “circulation form”**:

        * Radial-equilibrium ODE (Ciappi’s simplified outcome)

            Let $\Gamma(r) := 2\pi r,V_\theta(r)$. Then
            $$
            \frac{d}{dr}\Big(V_x^2\Big) = -\frac{1}{4\pi^2 r^2}\frac{d}{dr}\Big(\Gamma^2\Big).
            $$
            
            Equivalently,
            $$
            \frac{d}{dr}\Big(V_x^2 + V_\theta^2\Big) + \frac{2V_\theta^2}{r} = 0
            \quad\Longleftrightarrow\quad
            \frac{d(V_x^2)}{dr} = -\frac{1}{r^2}\frac{d}{dr}(r^2 V_\theta^2),
            $$
            and with $\Gamma=2\pi r V_\theta$ one recovers the first form.

        * Integration and closure

            Integrate once in $r$:
            $$
            V_x^2(r) = C + \text{(known integral of }V_\theta(r)\text{)}.
            $$
            The constant (C) is fixed by **global continuity**:
            $$
            Q_{\text{out}}=\int_{R_i}^{R} 2\pi r{}V_x(r){}dr=Q_{\text{in}}.
            $$
            
            **Summary Route A:** One **first-order ODE** in $r$, **no actuator disk jump modeled explicitly**, **constant total pressure** across the span downstream. Simple to implement; produces a mean-flow redistribution law $V_x(r)) given (V_\theta(r)$.

    * **ROUTE B — GATO (PERTURBATION + BESSEL EIGENMODES + ACTUATOR DISK)**

        * Decomposition into base + perturbation

            Introduce a **uniform base axial flow** $U$ and a **perturbation** $v_x'(r,x)$:
            $$
            V_x(r,x) = U + v_x'(r,x)\qquad V_\theta(r,x) \text{ given on } x=0^+ \text{ (from cascade)}.
            $$
            Assume the **meridional perturbation is irrotational** near the disk, so there exists a potential $\phi$ with $v_x'=\partial \phi/\partial x$.

        * Governing PDE for the perturbation (Laplace)

            For axisymmetric, incompressible, irrotational perturbation in $(r,x)$:
            $$
            \nabla^2 \phi = \frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial \phi}{\partial r}\right)
            +\frac{\partial^2 \phi}{\partial x^2}=0,
            \quad\Longrightarrow\quad
            \frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial v_x'}{\partial r}\right)
            +\frac{\partial^2 v_x'}{\partial x^2}=0.
            $$

        * Separation of variables and radial eigenproblem

            Seek $v_x'(r,x)=\sum_n A_n e^{-\lambda_n x} F_n(r)$. Then $F_n(r)$ solves
            $$
            \frac{1}{r}\frac{d}{dr}\left(r\frac{dF_n}{dr}\right) - \lambda_n^2 F_n = 0,
            \quad\Rightarrow\quad
            F_n(r)=J_0(\lambda_n r) + b_n Y_0(\lambda_n r).
            $$
            **Boundary conditions** at $r=R_i,R$ (e.g., Dirichlet $F_n=0$ or Neumann $F_n'=0)$ determine the **eigenvalues** $\lambda_n$ and the **mixing** $b_n$. Orthogonality holds with **weight $r$**.

        * Actuator disk boundary condition (radial Euler at $x=0^+$)

            At the disk plane, the **radial derivative of the axial perturbation** relates to swirl and the **radial derivative of the total-pressure drop**:
            $$
            G(r)=\left.\frac{\partial v_x'}{\partial r}\right|_{x=0^+}
            =\frac{1}{2V_x(r,0)}\frac{V\theta(r,0)}{r}\frac{d}{dr}\big(rV_\theta(r,0)\big)
            -\frac{1}{\rho}\frac{d}{dr}\Delta p_0(r),
            $$
            with the **sign convention** $\Delta p_0(r)=p_t(-0,r)-p_t(+0,r)>0$ for a drop (if your storage uses the negative of this, flip the sign of the derivative accordingly).

        * Projection to determine amplitudes

            Use orthogonality with weight $r$. Let $F_n'=\tfrac{dF_n}{dr}$. Then
            $$
            A_n =\frac{\displaystyle\int_{R_i}^{R} r{}G(r){}F_n'(r){}dr}
            {\displaystyle\int_{R_i}^{R} r{}\big(F_n'(r)\big)^2{}dr}.
            $$

        * Reconstruct the downstream jump and the downstream field
        
            At the disk plane,
            $$
            v_x'(r,0^+) = \sum_n A_n F_n(r)\qquad
            V_x(+0,r) = V_x(-0,r) - 2{}v_x'(r,0^+).
            $$
            Downstream decay for (x>0):
            $$
            v_x'(r,x) = \sum_n A_n e^{-\lambda_n x}F_n(r).
            $$

    * **SIDE-BY-SIDE: WHAT EXACTLY DIVERGES IN THE DERIVATIONS?**

        * **Unknown solved for**
        
          * **A:** $V_x(r)$ (mean outlet axial velocity).
          * **B:** $v_x'(r,x)$ (perturbation), then obtain $V_x(+0,r)=V_x(-0,r)-2v_x'(r,0^+)$.
        
        * **Pressure treatment**
        
          * **A:** impose $dp_t/dr=0$ at the plane → pressure term drops out; only swirl controls $d(V_x^2)/dr$.
          * **B:** keep $\Delta p_0(r)$ and its **radial derivative** $d\Delta p_0/dr$ as a forcing term that competes with swirl in setting $G(r)$.
        
        * **Mathematical structure**
        
          * **A:** first-order **ODE in $r$**; integrate and set a constant with continuity.
          * **B:** **eigenfunction expansion** in $r$ (Bessel), **exponential decay** in $x$, amplitudes from **projection** with the disk boundary condition.
        
        * **Continuity enforcement**
        
          * **A:** choose the integration constant so that $Q_{\text{out}}=Q_{\text{in}}$.
          * **B:** continuity is built into the disk relation $V_x(+0)=V_x(-0)-2v_x'$ and the modal reconstruction; globally one still checks $Q_{\text{out}}=Q_{\text{in}}$ as a consistency condition.
        
        * **Physical content**
        
          * **A:** “mean-only” redistribution consistent with **uniform total pressure** downstream; no explicit actuator model.
          * **B:** “perturbation + eigenmodes” with **actuator disk physics**: jump conditions at $x=0$, **spanwise BCs** at hub/tip, and **downstream decay**.

    * **MINIMAL “BRIDGE” BETWEEN BOTH VIEWS**

        If in Route B you set $d\Delta p_0/dr=0$ (spanwise-uniform disk loss) and take a single dominant mode, the projection reduces to “swirl-driven” shaping of $v_x'$ that, at the plane, behaves similarly to a **single-constant integration** in Route A. Conversely, Route A can be seen as the **mean-field limit** of Route B when the disk forcing is uniform and only the slowest radial structure survives.

---


**SUMMARY**

* **Ciappi**: ODE–based, assumes constant total pressure, no explicit actuator disk, simple but restrictive.
* **Gato**: Modal Bessel expansion, explicit actuator disk with pressure jump, more complex but physically richer.
