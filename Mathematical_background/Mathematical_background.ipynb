# SQUARE-INTEGRABILITY, HILBERT SPACES, AND TURBULENCE

---

**THE CORE QUESTION**

Why do we insist on **square-integrability** ($L^2$) when analyzing turbulence with Fourier methods, instead of cubic-integrability ($L^3$) or other $L^p$ spaces?

The answer lies in the difference between **Banach spaces** and **Hilbert spaces**, and in the physics of turbulence (energy is quadratic in velocity).

---

**VECTOR SPACES, NORMS, AND BANACH SPACES**

**Definition (Norm):**
A **norm** on a vector space $V$ is a function $\|\cdot\|:V\to\mathbb{R}$ satisfying, for all $f,g\in V$ and scalars $\alpha$:

1. $\|f\|\ge 0$, and $\|f\|=0$ iff $f=0$.
2. $\|\alpha f\|=|\alpha|\|f\|$.
3. $\|f+g\|\le \|f\|+\|g\|$ (triangle inequality).

**Definition (Banach space):**
A **Banach space** is a vector space with a norm, {ref}`complete <note_on_Cauchy_sequences>` in the metric induced by that norm. Completeness means: every Cauchy sequence in the space converges to an element of the space.

**Examples:**

* $L^1(\mathbb{R})$: functions with finite $\|f\|_1=\int |f(x)|dx$.
* $L^p(\mathbb{R})$ in general.

**Counterexample:**
The space of continuous functions $C[0,1]$ with the $L^1$-norm is *not complete*. A Cauchy sequence of continuous functions can converge to a discontinuous $L^1$ function that isn’t in $C[0,1]$. So $C[0,1]$ with this norm is not Banach.

---

**HILBERT SPACES: INNER PRODUCT GEOMETRY**

**Definition (Inner product):**
An **inner product** on a vector space $V$ is a function $\langle \cdot,\cdot\rangle:V\times V\to\mathbb{C}$ such that, for all $f,g,h\in V$, scalars $\alpha$:

1. $\langle f,g\rangle=\overline{\langle g,f\rangle}$.
2. $\langle \alpha f+g,h\rangle=\alpha\langle f,h\rangle+\langle g,h\rangle$.
3. $\langle f,f\rangle \ge 0$, with equality iff $f=0$.

The norm is induced by the inner product: $\|f\|=\sqrt{\langle f,f\rangle}$.

**Definition (Hilbert space):**
A **Hilbert space** is a complete inner product space.

**Example:**
$L^2(\mathbb{R})$ with

$$
\langle f,g\rangle=\int f(x)\,\overline{g(x)}\,dx.
$$

**Counterexample:**
$L^1(\mathbb{R})$ is not a Hilbert space because no inner product generates the $L^1$-norm. That is, you can’t find $\langle \cdot,\cdot\rangle$ with $\|f\|=\sqrt{\langle f,f\rangle}=\int |f|$.

---

**WHY HILBERT STRUCTURE IS SPECIAL**

The inner product in Hilbert spaces enables geometry:

* **Orthogonality:** $\langle f,g\rangle=0$.
* **Projection theorem:** closest point in a subspace is given by orthogonal projection.
* **Pythagoras:** $\|f+g\|^2=\|f\|^2+\|g\|^2$ if $f\perp g$.
* **Orthonormal bases:** generalization of perpendicular axes.

This structure **does not exist** in Banach spaces unless they are Hilbert.

**Counterexample (Banach but not Hilbert):**
In $L^1$, define $f(x)=1$ and $g(x)=-1$. Their “norm” relation

$$
\|f+g\|_1 = 0,\quad \|f\|_1+\|g\|_1=2,
$$

shows no inner product structure exists to explain this through orthogonality.

---

**FOURIER ANALYSIS AND $L^2$**

Why Fourier analysis naturally belongs to Hilbert spaces:

* Exponentials $e^{ikx}$ form an **orthonormal basis** in $L^2$.
* Fourier coefficients are **inner products**:

  $$
  \hat f(k)=\langle f, e^{ikx}\rangle.
  $$
* Plancherel’s theorem:

  $$
  \|f\|_2^2=\int |f(x)|^2dx=\int |\hat f(k)|^2 dk.
  $$

**Counterexample:**
In $L^1$, Fourier transforms exist, but no energy preservation: Parseval’s identity fails. Example:

* $f(x)=1/(1+x^2)$ is in $L^1$.
* Its Fourier transform is finite.
* But $\|f\|_1$ does not equal $\|\hat f\|_\infty$; no symmetry like in $L^2$.

---

**WHY “SQUARE” AND NOT “CUBIC”?**

* **Mathematical reason:** Only $L^2$ gives us an inner product, and with it, orthogonality and unitarity of Fourier transforms. $L^3$ or $L^4$ are Banach spaces without inner products.
* **Physical reason:** Turbulence is governed by kinetic energy, quadratic in velocity:

  $$
  E=\tfrac{1}{2}\rho u^2.
  $$

  Second moments (variance, covariance) matter because they define energy spectra. Third or fourth moments (skewness, flatness) have uses, but they do not define the energy distribution.

---

**CONNECTION TO TURBULENCE**

* If turbulence were globally in $L^2$, we’d have a well-defined Fourier transform and Parseval’s theorem.
* But turbulence is infinite in extent → $\int |u|^2=\infty$.
* What remains meaningful is the **spectral tensor** $\Phi_{ij}(\mathbf k)$: variance density in k-space.
* Mann’s Fourier–Stieltjes representation bypasses the lack of $L^2$ by defining turbulence directly via stochastic increments $dZ(\mathbf k)$ whose covariance is $\Phi_{ij}(\mathbf k)$.

So square-integrability is the natural mathematical structure, but turbulence falls just outside it — hence the stochastic generalization.

---

**DIAGRAMMATIC ANALOGY**

Visualize the difference between Hilbert and Banach spaces with geometry:

| Feature                | Hilbert space (e.g., $L^2$)                       | Banach space (e.g., $L^p, p\neq 2$) |
| ---------------------- | ------------------------------------------------- | ----------------------------------- |
| Structure              | Inner product + norm                              | Norm only                           |
| Norm origin            | From inner product                                | Defined directly                    |
| Angles / Orthogonality | Yes (via inner product)                           | No                                  |
| Projection theorem     | Holds                                             | Not available                       |
| Pythagoras’ theorem    | Holds ($\|f+g\|^2=\|f\|^2+\|g\|^2$) if $f\perp g$ | Not meaningful                      |
| Basis                  | Orthonormal bases (e.g. Fourier)                  | Bases may exist, but not orthogonal |
| Fourier transform      | Unitary (Plancherel, Parseval hold)               | Not unitary; energy identity lost   |
| Completeness           | Yes                                               | May or may not (Banach requires it) |

**Analogy:**


* A **Hilbert space** is like Euclidean geometry: you have a ruler **and** a protractor — you can measure both lengths and angles, decompose vectors into orthogonal axes.
* Banach space is like a space where you only have a ruler (to measure lengths), but no protractor (no angles). You can measure size, but not orthogonality.

---

**SUMMARY**
Square-integrability is special because $L^2$ is a Hilbert space. Only there do we have inner products, orthogonality, and unitary Fourier transforms. This matches physics where energy is quadratic in its hamiltonian parameters, like potential energy or kinetic energy.

**CAUCHY SEQUENCES**

**Definition (metric space).**
A sequence $(x_n)$ in a metric space $(X,d)$ is **Cauchy** if

$$
\forall \varepsilon > 0,\ \exists N \in \mathbb{N}\ \text{such that } m,n \geq N \implies d(x_m,x_n)<\varepsilon.
$$

Intuition: beyond some point, *all* terms of the sequence are arbitrarily close to each other, even if you don’t know the limit.

**In normed spaces:** $d(x,y)=\|x-y\|$.
**In Hilbert spaces:** the norm comes from the inner product: $\|f\|=\sqrt{\langle f,f\rangle}$.

---

**COMPLETENESS**

* A space is **complete** if every Cauchy sequence converges to an element *within the space*.
* A **Banach space** is a complete normed space.
* A **Hilbert space** is a complete inner-product space (and thus also a Banach space).

So completeness is the property that “no limits are missing.”

---

**EXAMPLES AND COUNTEREXAMPLES**

* **In $\mathbb{R}$:**

  * $x_n=1/n$ is Cauchy and converges to 0.
  * $x_n=(-1)^n$ is not Cauchy (oscillates forever).

* **In $\mathbb{Q}$:**

  * The sequence of decimal truncations to $\sqrt{2}$: $1.4,1.41,1.414,\dots$ is Cauchy, but has no limit in $\mathbb{Q}$. Shows incompleteness of rationals.

* **In $L^2[0,1]$:**

  * If $\|f_m-f_n\|_{L^2}\to 0$, then the sequence converges in $L^2$. This space is complete (Hilbert).

* **In $C[0,1]$ with $L^1$-norm:**

  * Take continuous approximations $f_n$ to the discontinuous step function $\chi_{[0,1/2]}$. The sequence is Cauchy in $L^1$ but its limit $\chi_{[0,1/2]}\notin C[0,1]$. So this space is **not** complete.

---

**WHY CAUCHY SEQUENCES MATTER FOR ANALYSIS**

* To **prove convergence** of a sequence, you can show it’s Cauchy (if the space is complete).
* To **disprove convergence**, find an $\varepsilon>0$ such that subsequences stay apart by at least $\varepsilon$.
* To **test completeness**, construct a Cauchy sequence that leaves the space (like the step function example).

# ON LEBESGUE AND STIELTJES INTEGRALS

**CONVENTIONAL LEBESGUE (OR RIEMANN) INTEGRAL**

* You integrate a function $f(x)$ *against* $dx$, i.e. the length of an infinitesimal interval.
* Visually: partition the $x$-axis into small bins $\Delta x$, multiply $f(x)\Delta x$, and sum.
* The “measure” here is just *length*.
* Formula:

  $$
  \int f(x)\,dx = \lim_{\Delta x\to 0}\sum f(x_i)\,\Delta x
  $$

**Sketch idea**:

* Draw $f(x)$ as a curve.
* On the $x$-axis, shade rectangles of base $\Delta x$, height $f(x)$.
* Area under curve = sum of these rectangles.

---

**STIELTJES INTEGRAL**

* Generalization: integrate *against* an arbitrary function $g(x)$, which defines a measure through its increments $\Delta g = g(x+\Delta x)-g(x)$.
* Formula:

  $$
  \int f(x)\,dg(x)=\lim_{\Delta x\to 0}\sum f(x_i)\,\Delta g(x_i).
  $$
* If $g(x)=x$, this reduces to the Lebesgue/Riemann integral.
* If $g(x)$ is *noisy* or *jumps*, then increments are irregular.

**Sketch idea**:

* Draw the same $f(x)$.
* Instead of uniform rectangles, you attach “weights” proportional to $\Delta g$, which may vary in size, or even flip sign.
* The “base” of the rectangles isn’t uniform length anymore; it’s dictated by how $g$ changes.

---

**STOCHASTIC FOURIER-STIELTJES INTEGRAL**

* Take a function $u(x)$ defined as a generalized Fourier-series:

  $$
  u(x) = \int e^{i k x}\,dZ(k),
  $$

  where $dZ(k)$ is a *stochastic measure increment* .
* Intuition: instead of summing waves with deterministic amplitudes $A(k)\,dk$, you sum with **random increments** $dZ(k)$.
* Each increment contributes a small random “kick” weighted by the oscillatory kernel $e^{i k x}$.

**Sketch idea**:

* Imagine the $k$-axis chopped into bins.
* For each bin, rather than deterministic “height × width”, you toss a dice (random increment $dZ$) with variance $\propto \frac{\langle dZ_i^*(\mathbf{k})\, dZ_j(\mathbf{k})\rangle}{dk_1\,dk_2\,dk_3}=\Phi_{ij}(\mathbf{k})$, with $\Phi_{ij}(\mathbf{k})$ being a rank-2 tensor that gives the variance density of the random increments in Fourier space.
* Add up all random phasors $e^{i k x}\,dZ(k)$.
* Averaging over realizations gives you a smooth overall quantity.

---

**VISUAL REPRESENTATION**

Here’s the full comparison, side by side:
![01](Figs\01_Lebesgue_Stieltjes_integrals.svg)

1. **Lebesgue / Riemann (left):**

   * Uniform partitions, all bases have equal $\Delta x$.
   * Contributions are simply $f(x)\,\Delta x$.

2. **Stieltjes deterministic (middle):**

   * Partitions distorted by a deterministic function $g(x)$.
   * Bases vary in width; the green dashed curve shows $g(x)$ driving the partition.
   * Contributions are $f(x)\,\Delta g(x)$.

3. **Stieltjes stochastic (right):**

   * Partitions are random (red dashed lines).
   * Weights (colored rectangles) vary randomly in size and even sign — like stochastic increments $dZ$.
   * This mimics Mann’s Fourier–Stieltjes: the measure is random, and the average over many realizations yields the spectral tensor.

This makes it clear:

* **Lebesgue:** uniform increments of length.
* **Deterministic Stieltjes:** nonuniform increments determined by a function.
* **Stochastic Stieltjes:** random increments, the setting for turbulence.

---

**SUMMARY**:

* **Lebesgue:** measure = length.
* **Stieltjes:** measure = increments of some function $g(x)$.
* **Stochastic Stieltjes:** measure = *random increments* $dZ(k)$ with covariance $\propto$ spectral tensor.

:::::{note} WHY LEBESGUE AND STIELTJES INTEGRALS ARE RELEVANT

**RIEMANN INTEGRAL: LIMITATIONS**

* **Definition (Riemann):** Partition the domain into subintervals of equal or variable width $\Delta x$. Approximate area by rectangles $f(x_i)\Delta x$. Limit as $\Delta x\to 0$ gives the integral.

  $$
  \int_a^b f(x)\,dx = \lim_{|\Delta x|\to 0}\sum f(x_i)\,\Delta x_i.
  $$

* **Limitations:**

  * Needs $f$ to be “nice” (few discontinuities, bounded).
  * Fails for many useful functions in analysis and probability (e.g., indicator functions on rationals, highly oscillatory functions).
  * Hard to handle integration with respect to measures other than “length”.

---

**LEBESGUE INTEGRAL: GENERALIZATION VIA MEASURE**

* **Key idea:** Instead of chopping up the **domain** (x-axis), chop up the **range** (y-axis) of the function.

* Define integrals via *measures* of sets, not rectangles.

* **Measure theory:**
  A *measure* $\mu$ assigns a nonnegative size to subsets (length, area, probability, etc.).
  The Lebesgue integral is defined as

  $$
  \int f\,d\mu
  $$

  where $\mu$ can be length, probability, surface measure, etc.

* **Formal construction:**

  1. Start with simple functions (step functions).
  2. Approximate any measurable $f$ by increasing sequence of simple functions.
  3. Define $\int f\,d\mu = \lim \sum c_i \mu(A_i)$.

* **Why relevant:**

  * Integrates *much larger class* of functions (bounded measurable, even some unbounded).
  * Works seamlessly with probability: expectations are Lebesgue integrals $\mathbb E[X]=\int x\,d\mathbb P$.
  * Foundation of Fourier analysis and Hilbert spaces ($L^2$).

---

**LEBESGUE–STIELTJES INTEGRAL: WEIGHTED MEASURES**

* **Motivation:** Sometimes, we don’t want to integrate wrt “length”, but wrt another cumulative function $g(x)$.

* **Definition:**

  $$
  \int f(x)\,dg(x).
  $$

  Here, $g(x)$ is monotone increasing; it defines a measure $\mu_g$ via

  $$
  \mu_g([a,b]) = g(b)-g(a).
  $$

  Then the Stieltjes integral is just the Lebesgue integral wrt $\mu_g$.

* **Examples:**

  * If $g(x)=x$, then $\mu_g$ is length → standard Lebesgue.
  * If $g(x)$ is a CDF of a random variable, then $\int f(x)\,dg(x)=\mathbb E[f(X)]$.
  * If $g$ has jumps, the integral automatically handles discrete contributions.

* **Relevance:** This makes Lebesgue integrals flexible — they can integrate functions wrt length, probability, distributions, or custom weights.

---

**STOCHASTIC (FOURIER–STIELTJES) INTEGRALS**

* **In Mann’s turbulence model:**
  $u(x)=\int e^{ikx}\,dZ(k).$
  Here $dZ(k)$ is *not a deterministic measure increment*, but a **random measure** with

  $$
  \mathbb E[dZ_i^*(\mathbf k)dZ_j(\mathbf k')] = \Phi_{ij}(\mathbf k)\,\delta(\mathbf k-\mathbf k')\,d^3k\,d^3k'.
  $$

* **Justification:**

  * We cannot represent turbulence velocities as Fourier transforms with deterministic coefficients (no square-integrability over infinite space).
  * By allowing random measure increments, we sidestep this: each increment is random, but its covariance structure reproduces the spectral tensor.
  * This is precisely a **stochastic Fourier–Stieltjes integral**.

* **Why relevant:**

  * Defines turbulence fields rigorously in the infinite-domain setting.
  * Connects covariance (physical space) and spectral tensor (wavenumber space).
  * Ensures realizability (positivity of spectral density).

---

**CONCEPTUAL CONTRAST**

* **Riemann:** intuitive, geometric “area under curve”, but limited.
* **Lebesgue:** powerful generalization, allows integration wrt arbitrary measures, foundation of probability.
* **Lebesgue–Stieltjes:** Area under curve wrt *another function’s increments* (flexible weighting). Adapts Lebesgue to arbitrary monotone functions, key for expectations wrt distributions.
* **Stochastic Stieltjes:** Area under curve wrt *random increments*, encoding randomness + spectral properties. Extends Lebesgue-Stieltjes to random measures.

---

**SUMMARY**

* **Lebesgue integrals**: unify measure and integration; essential for probability, Hilbert spaces, Fourier analysis.
* **Stieltjes integrals**: generalize to arbitrary measures (e.g. distribution functions).
* **Stochastic Stieltjes integrals**: generalize further to random measures, exactly what’s needed to represent turbulent velocity fields via their spectral tensor.

    (tip_interpreting_Lebesgue_integral)=
    ::::{tip} Interpreting the Lebesgue integral subplot

    **CONTRAST WITH CONVENTIONAL INTUITION ON INTEGRALS**

    * In **Lebesgue integration**, you don’t cut the $x$-axis into pieces (like Riemann).
    * You cut the **$y$-axis (the range of $f$)** into slices of height $\Delta f$.
    * For each slice $[y_i,y_{i+1}]$, you measure the **set of $x$-values** where $f(x)$ lies in that slice.
    * Then the contribution of that strip is:
    
      $$
      (y_{i+1}-y_i)\;\cdot\;\mu\big(\{x : f(x)\in [y_i,y_{i+1}]\}\big).
      $$
    
    So the horizontal “strips” are **not filled everywhere** across the domain, but only over those parts of $x$ where $f(x)$ is high enough to reach that level.

    **CORRECT MENTAL PICTURE**
    
    * **Riemann:** add up vertical *columns* of area.
    * **Lebesgue:** add up horizontal *layers* of area, but only where the curve actually extends above that height.
    
    So if a function dips below some value, the Lebesgue strip at that height simply has *zero width* in that region.

    (hint_Dirichlet_function)=
    :::{hint} Practical case with Dirichlet function

    **THE DIRICHLET FUNCTION**
    
    On the interval $[0,1]$, define
    
    $$
    f(x) = \begin{cases}
    1, & x \in \mathbb{Q}, \\
    0, & x \in \mathbb{R}\setminus\mathbb{Q}.
    \end{cases}
    $$
    
    * **Every subinterval** of $[0,1]$ contains rationals and irrationals (both are dense).
    * So $\sup f = 1$ and $\inf f = 0$ in every interval.
    * In Riemann integration, upper Darboux sums are always 1 and lower sums are always 0 → they never meet → **not Riemann integrable**.
    
    ---
    
    **THE PROVING ARGUMENT: RATIONALS HAVE LEBESQUE MEASURE ZERO**
    
    To evaluate the Lebesgue integral, we need to know the measure of the set $\{x:f(x)=1\}$, i.e. the rationals in $[0,1]$.
    
    * By definition, a set has Lebesgue measure 0 if, for every $\varepsilon>0$, you can cover it with a collection of intervals whose **total length < $\varepsilon$**.
    * For rationals:
    
      * List them (countable set).
      * Around each rational $q_n$, put an interval of length $\varepsilon/2^n$.
      * Total length $\leq \varepsilon$.
    * Since this works for *any* $\varepsilon>0$, the rationals have **measure zero**.
    
    This is the **vertical slicing / interval-cover argument**: it’s not the definition of the Lebesgue integral, but a tool to prove measure-zero.
    
    ---
    
    **COMPUTING OPERATION: LEBESQUE INTEGRAL VIA HORIZONTAL SCALING**
    
    The Lebesgue integral is defined as
    
    $$
    \int_0^1 f(x)\,dx = \int_0^\infty \mu\{x: f(x) > t\}\,dt.
    $$
    
    Now compute:
    
    * For $t \in (0,1]$:
      $\{x:f(x)>t\} = \{x:f(x)=1\} = \text{rationals}$.
      This set has measure 0.
    
    * For $t \leq 0$:
      $\{x:f(x)>t\} = [0,1]$, measure 1.
      But this doesn’t affect the integral, because we integrate $t$ only over $t>0$.
    
    So
    
    $$
    \int_0^1 f(x)\,dx = \int_0^1 0\,dt = 0.
    $$
    
    This is the **horizontal slicing picture**: chop the **y-axis** into bands, ask “how wide is the set of $x$ where $f(x)$ exceeds this level?”, then integrate those widths. For the Dirichlet function, those widths are 0 for all $t>0$.

(note_Dirichlet_integrability)=
:::::{note} Dirichlet integrability

**WHY DIRICHLET IS NOT RIEMANN-INTEGRABLE, BUT IS LEBESGUE-INTEGRABLE**

* **Riemann integration** is based on partitioning the *x-axis* into small intervals and using sup/inf of the function inside each interval.

* For the Dirichlet function

  $$
  f(x) = \begin{cases}
  1, & x \in \mathbb{Q} \cap [0,1], \\
  0, & x \in \mathbb{R}\setminus \mathbb{Q},
  \end{cases}
  $$

  every subinterval contains rationals (value 1) and irrationals (value 0).
  → Sup = 1, inf = 0, so upper Darboux sums = 1, lower = 0 for *all partitions*.
  → No limit = **not Riemann integrable**.

* **Lebesgue integration** instead asks: *for each horizontal slice of function values, what is the measure of the set of $x$ that fall there?*

  * For any $t>0$, the set $\{x:f(x)>t\}$ = rationals, which have **measure zero**.
  * So the Lebesgue integral is $\int_0^1 0 \, dt = 0$.
    → **Lebesgue integrable with value 0.**

**The key insight:** Lebesgue is better adapted to “pathological” functions that are *very discontinuous*, because it integrates with respect to *measure*, not by “averaging” over intervals.

---

**IS THIS AN ODDITY?**

Yes and no:

* The Dirichlet function is a **classic teaching example** because it is extreme: discontinuous everywhere, “spiky.” It’s not something you’d find in physics.
* But the underlying issue (Riemann failing where Lebesgue works) is **not rare** in analysis. Any function with “too many discontinuities” can break Riemann but survive Lebesgue.

---

**PHYSICALLY RELEVANT EXAMPLES**

This is where it gets interesting. You’re right to want a function that’s **not just a pathological toy**. A couple of cases:

* 1. Indicator functions of fractal sets

    * Consider the Cantor set $C \subset [0,1]$.
    * Define $f(x)=1$ if $x \in C$, else $0$.
    * This is **not Riemann integrable**, because $C$ is uncountable, dense in itself, has measure zero, and the set of discontinuities has measure > 0.
    * But in Lebesgue’s sense:
      $\int f = \mu(C)=0$.
    * Fractals show up in physics (porous media, turbulence scaling, strange attractors).

* 2. Highly oscillatory functions

    * Example: $f(x) = \sin(1/x)$ for $x>0$, $f(0)=0$.
    * On $[0,1]$, $f$ is Riemann integrable.
    * But consider variants like
    
      $$
      g(x)=\begin{cases}
      1,& x \in \mathbb{Q}, \\
      \sin(1/x), & x \notin \mathbb{Q},
      \end{cases}
      $$
    
      which blends the Dirichlet idea with oscillation. Riemann fails; Lebesgue succeeds.

* 3. Discontinuous signals / impulses

    * The **Dirac delta** $\delta(x)$ is not Riemann integrable (not even a function in the usual sense), but in Lebesgue/distribution theory it can be treated rigorously.
    * In physics, delta-like distributions are *everywhere* (point charges, point forces, Green’s functions).

---

**PHYSICAL TAKEAWAY**

* **Riemann** is fine for “nice” functions: continuous, piecewise continuous, bounded with only finitely many discontinuities.
* **Lebesgue** is needed once you deal with:

  * “Wild” sets (fractals, rationals).
  * Probability theory (random variables often live naturally in $L^2$, the Lebesgue square-integrable world).
  * Physics distributions (delta, white noise, turbulence spectra).

So: while the **Dirichlet function** itself is a mathematical curiosity, the **need for Lebesgue integration** is deeply physical: turbulence energy spectra, random processes, probability densities, and even just handling the delta function all live in Lebesgue’s framework.

::::{tip} On Darboux sums

**IDEA OF DARBOUX SUMS**

Given a bounded function $f:[a,b]\to\mathbb R$:

1. **Partition the interval** into subintervals:

   $$
   P = \{a = x_0 < x_1 < x_2 < \dots < x_n = b\}.
   $$

2. On each subinterval $[x_{i-1}, x_i]$:

   * Let $m_i = \inf\{f(x): x \in [x_{i-1},x_i]\}$ (the lowest value of $f$ there).
   * Let $M_i = \sup\{f(x): x \in [x_{i-1},x_i]\}$ (the highest value of $f$ there).

3. Define the **lower Darboux sum**:

   $$
   L(f,P) = \sum_{i=1}^n m_i \cdot (x_i - x_{i-1}).
   $$

   Define the **upper Darboux sum**:

   $$
   U(f,P) = \sum_{i=1}^n M_i \cdot (x_i - x_{i-1}).
   $$

So:

* $L(f,P)$ is like approximating the integral using the **lowest rectangle** in each subinterval.
* $U(f,P)$ uses the **highest rectangle** in each subinterval.

---

**CRITERION FOR RIEMANN INTEGRABILITY**

* The function $f$ is **Riemann integrable** on $[a,b]$ if and only if:

  $$
  \sup_{P} L(f,P) = \inf_{P} U(f,P).
  $$
* This common value is the Riemann integral $\int_a^b f(x)\,dx$.

In words: **if refining partitions makes the lower and upper sums squeeze together**, then the integral exists.

---

**WHY DIRICHLET FAILS**

* For the Dirichlet function, every subinterval contains rationals (value 1) and irrationals (value 0).
* So: $m_i = 0$, $M_i = 1$ for all intervals.
* Thus: $L(f,P)=0$, $U(f,P)=b-a$.
* No matter how fine the partition, they never converge → **not integrable**.

# ON COMMUTATORS

**THE DEFINITION**

Given two linear operators $A$ and $B$ acting on some space of functions (or vectors), their **commutator** is defined as

$$
[A,B] \;\equiv\; A B - B A.
$$

It measures the **extent to which the order of applying the operators matters**.

* If $[A,B]=0$, then $A$ and $B$ **commute**: $ABf = BAf$ for all functions $f$.
* If $[A,B]\neq 0$, then the operators **do not commute**: applying them in different orders produces different results.

---

**SIMPLE EXAMPLES**

* **PARTIAL DERIVATIVES IN DIFFERENT DIRECTIONS**

   $$
   [\partial_x,\partial_y]f = \partial_x(\partial_y f) - \partial_y(\partial_x f) = 0
   $$

   (for smooth functions). So mixed derivatives commute.
<br/>
<br/>
* **MULTIPLICATION VS DIFFERENTIATION**

   Let $A = x$ (multiplication by $x$), $B = \partial_x$. Then

   $$
   [x,\partial_x]f = x\partial_x f - \partial_x(x f) = x f' - (f + x f') = -f.
   $$

   So $[x,\partial_x] = -\,I$.
   Here, multiplication by $x$ and differentiation w\.r.t. $x$ don’t commute.
<br/>
<br/>
* **AN EASY GEOMETRIC PICTURE OF COMMUTATORS**

    Take two vector fields $X$ and $Y$ on a smooth manifold (think of them as differential operators: $X = a(x,y)\partial_x + b(x,y)\partial_y$, etc.).
    Each vector field generates a **flow**: moving a point a small step along $X$, or along $Y$.
    
    * Apply flow $X$ for a short time $\epsilon$, then flow $Y$ for the same time: you reach some new point.
    * Do it in reverse order (first $Y$, then $X$): you reach another point.
    
    If $X$ and $Y$ commute, both paths lead to the same point. If they don’t, there is a **leftover displacement** after going around the small “parallelogram” of flows.
    
    * As a visual intuition:
    
        * Imagine pushing a fluid parcel first along streamlines of $X$, then along $Y$.
        * Reversing the order yields a different final position if the vector fields twist relative to each other.
        * The difference is exactly the action of the commutator.

    ![05](Figs/05_commutator_parallelogram.gif)
 
    * We compare the action of two flows (operators) $X$ and $Y$ on a point in the plane.

        * **Flow $X$:** horizontal shift
        
          $$
          X: (x,y) \mapsto (x+t,\, y)
          $$

          — this just translates to the right by an amount $t$.

        * **Flow $Y$:** vertical shear depending on $x$
        
          $$
          Y: (x,y) \mapsto (x,\, y+t\,x).
          $$
        
          — the vertical displacement is proportional to the current $x$-coordinate.

      We start at the origin $(0,0)$ and apply each flow with the same small parameter $\varepsilon$.

        * **Path A (blue) = Apply $X$ then $Y$**

            1. Apply $X$: $(0,0) \mapsto (\varepsilon, 0)$.
            2. Then apply $Y$: vertical shift by $\varepsilon \cdot x = \varepsilon \cdot \varepsilon = \varepsilon^2$.
               Final point: $(\varepsilon, \varepsilon^2)$.

        * **Path B (red) = Apply $Y$ then $X$**

            1. Apply $Y$: but at the origin $(0,0)$, the vertical displacement is proportional to $x=0$.
               So **nothing happens**: $(0,0) \mapsto (0,0)$.
            2. Then apply $X$: $(0,0) \mapsto (\varepsilon,0)$.
               Final point: $(\varepsilon,0)$.

        * The gap (green arrow in the GIF)

            * Path A ends at $(\varepsilon, \varepsilon^2)$.
            * Path B ends at $(\varepsilon, 0)$.
            * The gap between them is $(0, \varepsilon^2)$.
            
          This leftover displacement is proportional to $\varepsilon^2 [X,Y]$, where $[X,Y]$ is the **commutator** of the two flows.

        * The meaning

            * If the two flows commuted, the blue and red paths would end at the same point: the “parallelogram” would close.
            * But they don’t: the parallelogram fails to close by a small displacement proportional to the commutator.
            * **That failure to close is the geometric meaning of the commutator.**

:::{admonition} On the particular calculation of the figure's commutator
:class: note

**VECTOR FIELDS (INFINITESIMAL GENERATORS)**

* $X$ = horizontal translation ⇒ $X = \partial_x$.
* $Y$ = vertical shear proportional to $x$ ⇒ $Y = x\,\partial_y$.

**COMMUTATOR CALCULATION**

* For any smooth $f(x,y)$,

    $$
    [X,Y]f \;=\; X(Yf)-Y(Xf)
    = \partial_x\!\big(x\,\partial_y f\big) \;-\; x\,\partial_y(\partial_x f).
    $$

* Compute:

    $$
    \partial_x\!\big(x\,\partial_y f\big) = (\partial_x x)\,\partial_y f + x\,\partial_{xy} f
    = \partial_y f + x\,\partial_{xy} f,
    $$
    
    $$
    x\,\partial_y(\partial_x f) = x\,\partial_{yx} f.
    $$

* Since mixed partials commute, $\partial_{xy} f=\partial_{yx} f$. Hence

    $$
    [X,Y]f \;=\; \partial_y f.
    $$
    
    So
    
    $$
    \boxed{[X,Y] \;=\; \partial_y,}
    $$

    i.e., the commutator is a **pure upward** vector field (unit speed in $y$).

**MATCH WITH THE ENDPOINTS IN THE FIGURE**

* With step size $\varepsilon$:

    * Path A ( $X$ then $Y$ ) from $(0,0)$:
    
      $$
      (0,0)\xrightarrow{X} (\varepsilon,0)\xrightarrow{Y} (\varepsilon,\, \varepsilon\cdot \varepsilon)= (\varepsilon,\,\varepsilon^2).
      $$
      
    * Path B ( $Y$ then $X$ ) from $(0,0)$:
    
      $$
      (0,0)\xrightarrow{Y} (0,0)\xrightarrow{X} (\varepsilon,0).
      $$
      
    * Difference (final minus final):
    
      $$
      (\varepsilon,\,\varepsilon^2) - (\varepsilon,\,0) = (0,\,\varepsilon^2).
      $$

    This is exactly $\varepsilon^2$ times the commutator vector field $[X,Y]=\partial_y$ evaluated at the start point—i.e., a small **vertical** displacement of size $\varepsilon^2$.

**GENERAL INTERPRETATION**

* **Mathematically**

    The commutator encodes the **non-compatibility** of two operations. It’s not just a technicality: the set of operators with the commutator as a binary operation forms a **Lie algebra**.

  In particular:

    * The commutator is **bilinear**: $[aA+bB,C] = a[A,C] + b[B,C]$.
    * It is **antisymmetric**: $[A,B] = -[B,A]$.
    * It satisfies the **Jacobi identity**:
    
    $$
    [A,[B,C]] + [B,[C,A]] + [C,[A,B]] = 0.
    $$
    
  Thus:
    
    * The set of differential operators with $[A,B]$ as defined forms a Lie algebra.
    * More generally, **any algebra of operators with the commutator bracket is a Lie algebra**.

* **Link to Lie groups**

    Lie algebras are the **infinitesimal structure** of Lie groups.
    
    * A **Lie group** is a continuous symmetry group (e.g. rotations, translations, Lorentz transformations).
    * Its **Lie algebra** is the tangent space at the identity, with the Lie bracket describing how infinitesimal symmetries combine.
    
    For example:
    
    * The group $SO(3)$ of 3D rotations has Lie algebra $\mathfrak{so}(3)$, consisting of $3\times3$ skew-symmetric matrices.
    * Their commutator is matrix commutator $[A,B]=AB-BA$.
    * This matches exactly the cross product in $\mathbb{R}^3$:
    
      $$
      [J_x,J_y] = J_z, \quad [J_y,J_z]=J_x, \quad [J_z,J_x]=J_y,
      $$
    
      where $J_i$ are infinitesimal generators of rotations.
    
    So: commutators of operators give rise to the algebraic skeleton of symmetries.

---

**PHYSICAL ITNERPRETATIONS ACROSS SEVERAL DISCIPLINES**

* **Quantum mechanics:**
  Observables correspond to operators on a Hilbert space. Their commutator measures whether two quantities can be measured simultaneously. For instance,

  $$
  [x, p_x] = i\hbar
  $$

  expresses the Heisenberg uncertainty principle: position and momentum cannot both be sharply defined.

* **Classical mechanics (Hamiltonian systems):**
  The commutator of operators corresponds to the **Poisson bracket** of classical observables. This structure encodes the dynamical evolution of phase-space functions.

* **Differential geometry:**
  If $A,B$ are vector fields, $[A,B]$ is their **Lie bracket**. It measures how much the flows generated by $A$ and $B$ fail to commute. This is central to the notion of curvature and integrability of distributions.

* **Fluid dynamics:**
  When operators involve material derivatives and spatial gradients, nonzero commutators reflect the fact that different derivatives (say, along shear vs across shear) don’t commute. Physically, this expresses shear coupling: neighboring fluid parcels advected differently distort gradients in a way that depends on order.

* **Control theory / robotics:**
  The commutator of control vector fields describes **reachable motions**: by alternating two control inputs in sequence, you can achieve motion in a new direction (nonholonomic effects, like parallel parking a car).

---

**SUMMARY**

A commutator is the algebraic measure of non-commutativity, its mathematical structure underlies **Lie algebras**, and physically it often encodes **incompatibility of operations, uncertainty, curvature, or new directions of motion**.
